## 3.

"Genere una visualización que muestre claramente la distribución horaria de accidentes en función del
tipo de vehículo involucrado (utilice la columna Vehicle Type). Comente brevemente si existe un
patrón claro que relacione la hora del día con ciertos tipos de vehículos."

Necesitamos unir ambos DataFrames por `Accident_Index`, y extraer la hora del accidente desde la columna `Time` en el archivo `accidentes.csv`.

````python
acc_veh = pd.merge(vehiculos, accidentes[['Accident_Index', 'Time']], on='Accident_Index', how='inner')
acc_veh.head(10)
````

````python
acc_veh['Hour'] = pd.to_datetime(acc_veh['Time'], errors='coerce').dt.hour.astype('Int64')
display(acc_veh[['Time', 'Hour']])
````

Contamos cuantos accidentes hay cada hora y por tipo de vehículo.

````python
conteo = acc_veh.groupby(['Hour', 'Vehicle_Type']).size().reset_index(name='count')
conteo
````

Calculamos la cantidad de accidentes por hora y luego la proporción de cada hora.

````python
total_por_hora = conteo.groupby('Hour')['count'].sum().reset_index(name='total')
display(total_por_hora)

conteo = conteo.merge(total_por_hora, on='Hour')
conteo['proporcion'] = conteo['count'] / conteo['total']
display(conteo)
````

Filtramos los accidentes con los primeros 5 tipos de vehiculos más involucrados.

````python
top_vehiculos = acc_veh['Vehicle_Type'].value_counts().head(5).index
acc_veh_top = acc_veh[acc_veh['Vehicle_Type'].isin(top_vehiculos)]
acc_veh_top.head(10)
````

````python
conteo_filtrado = conteo[conteo['Vehicle_Type'].isin(top_vehiculos)]
conteo_filtrado
````

````python
plt.figure(figsize=(12,6))
sns.lineplot(data=conteo_filtrado, x='Hour', y='proporcion', hue='Vehicle_Type', marker='o')
plt.title('Proporción de cada tipo de vehículo en los accidentes por hora')
plt.xlabel('Hora del día')
plt.ylabel('Proporción dentro de cada hora')
plt.grid(True)
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
````

## 4.

"Calcule un índice numérico que mida la severidad promedio de los accidentes por cada autoridad local
(Local Authority (District)), considerando en esto además el número de vehículos involucrados y
sus características. Presente las 10 autoridades locales con mayor severidad promedio.
"

Calculamos el numero de vehículos por accidente.

````python
vehiculos_por_accidente = vehiculos.groupby('Accident_Index').size().reset_index(name='num_vehiculos')
vehiculos_por_accidente
````

Juntamos con las columnas del archivo `accidentes` que nos dan informacion sobre la severidad y la autoridad local.

````python
acc_severidad = accidentes[['Accident_Index', 'Accident_Severity', 'Local_Authority_(District)']]
acc_sev_veh = pd.merge(acc_severidad, vehiculos_por_accidente, on='Accident_Index')
acc_sev_veh
````

````python
acc_sev_veh['Accident_Severity'].value_counts()
````

Implementamos el siguiente codigo para asignar un valor numerico al tipo de severidad.

````python
for sev in acc_sev_veh['Accident_Severity']:
  if sev == 'Slight':
    acc_sev_veh['Accident_Severity'] = acc_sev_veh['Accident_Severity'].replace(sev, 1)
  elif sev == 'Serious':
    acc_sev_veh['Accident_Severity'] = acc_sev_veh['Accident_Severity'].replace(sev, 2)
  else:
    acc_sev_veh['Accident_Severity'] = acc_sev_veh['Accident_Severity'].replace(sev, 3)
````

````python
acc_sev_veh['Accident_Severity'].value_counts()
````

Creamos el indice de severidad.

````python
acc_sev_veh['indice_severidad'] = acc_sev_veh['Accident_Severity'] * acc_sev_veh['num_vehiculos']
````

Realizamos el ranking de las 10 autoridades locales con mayor severidad promedio.

````python
ranking = acc_sev_veh.groupby('Local_Authority_(District)')['indice_severidad'].mean().sort_values(ascending=False).head(10)
print("Top 10 autoridades locales con mayor severidad promedio:")
print(display(ranking))
````
