#### Posibles llaves primarias y foráneas

El campo vin es el principal candidato a identificar vehículos, ya que corresponde al Vehicle Identification Number. Sin embargo, como vin no es único a nivel de fila en el CSV, no debería usarse como llave primaria de la tabla original completa. 

Una interpretación razonable es usar vin como llave primaria de una futura tabla Vehiculos, y usarlo como llave foránea en una tabla Ventas

En el caso de foráneas, 

make, podría usarse desde modelos o vehículos hacia fabricantes

model / trim, desde vehículos hacia modelos

state, desde ventas o vehículos hacia estados

seller, desde ventas hacia vendedores

vin, desde ventas hacia vehículos

### Mision 2

En base al enunciado y lo visto anteriormente, se pueden identificar las siguientes entidades:

#### 1. Fabricantes

Representa la marca o fabricante del vehículo.

**Atributos:**
id_fabricante y nombre

**Relación:**

Un fabricante puede tener muchos modelos, pero cada modelo pertenece a un solo fabricante.

Cardinalidad:

```text
Fabricantes 1 - N Modelos
```


#### 2. Modelos

Representa el modelo específico de un vehículo, asociado a un fabricante.

**Atributos:**

id_modelo, nombre, trim, body, id_fabricante


**Relación:**

Un modelo pertenece a un fabricante. Un mismo modelo puede estar asociado a muchos vehículos.

Cardinalidad:

```text
Fabricantes 1 - N Modelos
Modelos 1 - N Vehiculos
```


#### 3. Vehículos

Representa cada vehículo individual identificado por su VIN.

**Atributos:**
vin, year, transmission, color, interior, id_modelo

**Relación:**

Cada vehículo corresponde a un modelo. Un vehículo puede aparecer en una o más ventas registradas.

Cardinalidad:

```text
Modelos 1 - N Vehiculos
Vehiculos 1 - N Ventas
```

#### 4. Estados

Representa los estados de Estados Unidos que aparecen en la columna `state`.

**Atributos:**

id_estado, nombre_estado

**Relación:**

Un estado puede estar asociado a muchas ventas. En este modelo, `state` se asocia a la venta u observación, porque en los datos un mismo `vin` puede aparecer vinculado a distintos estados.

Cardinalidad:

```text
Estados 1 - N Ventas
```


#### 5. Vendedores

Representa a las entidades o personas que venden los vehículos.

**Atributos:**
id_vendedor, nombre

**Relación:**

Un vendedor puede participar en muchas ventas. Cada venta está asociada a un vendedor.

Cardinalidad:

```text
Vendedores 1 - N Ventas
```


#### 6. Ventas

Representa cada transacción u observación de venta registrada en el CSV.

**Atributos:**
id_venta, vin, id_vendedor, id_estado, condition, odometer, mmr, sellingprice, saledate

**Relación:**

Cada venta corresponde a un vehículo, un vendedor y un estado. Un vehículo puede tener más de una venta registrada, especialmente porque se observaron valores repetidos de `vin` en el CSV.

Cardinalidad:

```text
Vehiculos 1 - N Ventas
Vendedores 1 - N Ventas
Estados 1 - N Ventas
```

### Misión 3: Creación de tablas

Se crea una base de datos relacional llamada `vehicles.db`. La base contiene las tablas `Fabricantes`, `Modelos`, `Vehiculos`, `Estados`, `Vendedores` y `Ventas`.

Cada tabla posee una llave primaria. Las relaciones entre entidades se implementan mediante llaves foráneas. En particular, `vin` se usa como identificador único de cada vehículo en la tabla `Vehiculos`, mientras que la tabla `Ventas` usa un identificador propio `id_venta`, ya que un mismo vehículo puede aparecer en más de una venta.

Además, se activan las restricciones de llaves foráneas con `PRAGMA foreign_keys = ON` (en SQLite estas restricciones deben habilitarse explícitamente).

````python
import sqlite3

# crear la base de datos y establecer conexión
connection = sqlite3.connect("vehicles.db")
cursor = connection.cursor()

# activamos verificación de llaves foráneas
cursor.execute("PRAGMA foreign_keys = ON")
````

````python
# crear tablas
cursor.executescript("""
CREATE TABLE IF NOT EXISTS Fabricantes (
    id_fabricante INTEGER PRIMARY KEY AUTOINCREMENT,
    nombre TEXT NOT NULL UNIQUE
);

CREATE TABLE IF NOT EXISTS Modelos (
    id_modelo INTEGER PRIMARY KEY AUTOINCREMENT,
    nombre TEXT,
    trim TEXT,
    body TEXT,
    id_fabricante INTEGER,
    
    FOREIGN KEY (id_fabricante)
        REFERENCES Fabricantes(id_fabricante),
    
    UNIQUE(nombre, trim, body, id_fabricante)
);

CREATE TABLE IF NOT EXISTS Vehiculos (
    vin TEXT PRIMARY KEY,
    year INTEGER,
    transmission TEXT,
    color TEXT,
    interior TEXT,
    id_modelo INTEGER,
    
    FOREIGN KEY (id_modelo)
        REFERENCES Modelos(id_modelo)
);

CREATE TABLE IF NOT EXISTS Estados (
    id_estado INTEGER PRIMARY KEY AUTOINCREMENT,
    nombre_estado TEXT NOT NULL UNIQUE
);

CREATE TABLE IF NOT EXISTS Vendedores (
    id_vendedor INTEGER PRIMARY KEY AUTOINCREMENT,
    nombre TEXT NOT NULL UNIQUE
);

CREATE TABLE IF NOT EXISTS   Ventas (
    id_venta INTEGER PRIMARY KEY AUTOINCREMENT,
    vin TEXT NOT NULL,
    id_vendedor INTEGER NOT NULL,
    id_estado INTEGER NOT NULL,
    condition REAL,
    odometer REAL,
    mmr REAL NOT NULL,
    sellingprice REAL NOT NULL,
    saledate TEXT NOT NULL,
    
    FOREIGN KEY (vin)
        REFERENCES Vehiculos(vin),
    
    FOREIGN KEY (id_vendedor)
        REFERENCES Vendedores(id_vendedor),
    
    FOREIGN KEY (id_estado)
        REFERENCES Estados(id_estado)
);
""")

connection.commit()
````

````python
# ver tablas creadas
cursor.execute("""
SELECT name
FROM sqlite_master
WHERE type = 'table'
ORDER BY name;
""")

tablas = cursor.fetchall()
tablas
````

Detalles clave a considerar sobre el formato de salida: 

- fetchall() devolverá una lista de tuplas, por ejemplo: [('usuarios',), ('pedidos',)].
- Tablas del sistema: Si usa una base de datos con tablas internas del sistema (como sqlite_sequence), estas también aparecerán en la lista

### Misión 4

Para cargar los datos en la base relacional, se respeta el orden de dependencias definido en el modelo (si no no funcionará)

Primero se cargan las tablas que no dependen de otras tablas: `Fabricantes`, `Estados` y `Vendedores`. 

Luego se cargan `Modelos`, que depende de `Fabricantes`. 

Después se cargan `Vehiculos`, que depende de `Modelos`. Finalmente se carga `Ventas`, que depende de `Vehiculos`, `Vendedores` y `Estados`.

Como el campo `vin` identifica vehículos, pero no identifica de forma única cada fila del CSV, la tabla `Vehiculos` se carga usando una sola fila por cada `vin`. En cambio, la tabla `Ventas` mantiene todas las filas originales, ya que cada fila representa una observación o transacción de venta.

Para evitar problemas con valores faltantes en campos categóricos, estos se reemplazan por `"Desconocido"` antes de poblar las tablas. Los valores numéricos faltantes se mantienen como valores nulos.

````python
# copia del dataframe original para preparar la carga
df_carga = df.copy()

# columnas categóricas/textuales
cols_texto = [
    "make", "model", "trim", "body", "transmission",
    "vin", "state", "color", "interior", "seller", "saledate"
]

# reemplazar nulos categoricos por "Desconocido"
for col in cols_texto:
    df_carga[col] = df_carga[col].fillna("Desconocido").astype(str).str.strip()

# si algún string quedo vacio, también se reemplaza
df_carga[cols_texto] = df_carga[cols_texto].replace("", "Desconocido")
````

````python
# limpiamos tablas antes de cargar, para que si ejecutan de nuevo esto, funcione todo.

# se eliminan primero las tablas dependientes
cursor.executescript("""
DELETE FROM Ventas;
DELETE FROM Vehiculos;
DELETE FROM Modelos;
DELETE FROM Fabricantes;
DELETE FROM Estados;
DELETE FROM Vendedores;
""")

# reiniciar contadores autoincrementales
cursor.executescript("""
DELETE FROM sqlite_sequence
WHERE name IN (
    'Ventas',
    'Modelos',
    'Fabricantes',
    'Estados',
    'Vendedores'
);
""")

connection.commit()
````

Ahora, traducir los textos del CSV a los IDs reales de las tablas. Partamos por Fabricantes

````python
# obtenemos todos los fabricantes desde la tabla Fabricantes
cursor.execute("SELECT id_fabricante, nombre FROM Fabricantes")

# guardar el resultado de la consulta
filas_fabricantes = cursor.fetchall()
filas_fabricantes[:10]
````

````python
map_fabricantes = {}

# recorrer cada fila obtenida desde la base de datos
for fila in filas_fabricantes:
    id_fabricante = fila[0]
    nombre = fila[1]

    # asociamos el nombre del fabricante con su id
    map_fabricantes[nombre] = id_fabricante
````

Ahora se repite el proceso para el estado y vendedor. Se puede hacer más compacto.

````python
# mapa estado: nombre_estado -> id_estado
cursor.execute("SELECT id_estado, nombre_estado FROM Estados")
map_estados = {
    nombre_estado: id_estado
    for id_estado, nombre_estado in cursor.fetchall()
}

# mapa vendedor: nombre -> id_vendedor
cursor.execute("SELECT id_vendedor, nombre FROM Vendedores")
map_vendedores = {
    nombre: id_vendedor
    for id_vendedor, nombre in cursor.fetchall()
}
````

Ahora cargamos Modelos, que queda definido por fabricante, nombre de modelo, trim y body

````python
modelos_df = (
    df_carga[["make", "model", "trim", "body"]].drop_duplicates().copy()
)
````

````python
modelos_df.head()
````

````python
modelos = []

for _, row in modelos_df.iterrows():
    id_fabricante = map_fabricantes[row["make"]]
    modelos.append((
        row["model"],
        row["trim"],
        row["body"],
        id_fabricante
    ))

cursor.executemany("""
    INSERT OR IGNORE INTO Modelos(nombre, trim, body, id_fabricante)
    VALUES (?, ?, ?, ?)
""", modelos)

connection.commit()
````

````python
cursor.execute("""
SELECT id_modelo, nombre, trim, body, id_fabricante
FROM Modelos
""")

map_modelos = {
    (nombre, trim, body, id_fabricante): id_modelo
    for id_modelo, nombre, trim, body, id_fabricante in cursor.fetchall()
}
````

````python
map_modelos.keys()
````

Ahora vamos con Vehículos, que queda definido por vin, year, transmission, color, interior y modelo

````python
vehiculos_df = (
    df_carga.drop_duplicates(subset=["vin"], keep="first").copy()
)
# aqui hay sutilezas importantes. Hay vehículos que aparecen más de una vez según lo que vimos en la misión 1
# pero en la tabla Vehiculos, vin es la llave primaria, así que no podemos insertar dos veces el mismo vin.
# por eso, usamos keep="first" para quedarnos con la primera aparición de cada vin, y eliminar las filas duplicadas posteriores.

vehiculos = []

for _, row in vehiculos_df.iterrows():
    id_fabricante = map_fabricantes[row["make"]]
    id_modelo = map_modelos[
        (row["model"], row["trim"], row["body"], id_fabricante)
    ]
    
    vehiculos.append((
        row["vin"],
        int(row["year"]),
        row["transmission"],
        row["color"],
        row["interior"],
        id_modelo
    ))

cursor.executemany("""
    INSERT OR IGNORE INTO Vehiculos(
        vin,
        year,
        transmission,
        color,
        interior,
        id_modelo
    )
    VALUES (?, ?, ?, ?, ?, ?)
""", vehiculos)

connection.commit()
````

Como se ve, en esta etapa se carga la tabla `Vehiculos`. Dado que el campo `vin` identifica conceptualmente a cada vehículo, pero en el CSV existen algunos `vin` repetidos, se utiliza `drop_duplicates(subset=["vin"], keep="first")` para conservar una sola fila por vehículo. De esta forma, la tabla `Vehiculos` queda con un registro por cada `vin` único.

Esto permite separar correctamente la información del vehículo de la información de venta. Los datos que pueden cambiar entre una observación y otra, como `condition`, `odometer`, `seller`, `mmr`, `sellingprice` y `saledate`, no se cargan en `Vehiculos`, sino que se guardan posteriormente en la tabla `Ventas`. 

En cambio, en `Vehiculos` se almacenan atributos más propios del vehículo, como `vin`, `year`, `transmission`, `color`, `interior` y la referencia al modelo.

Para insertar cada vehículo, primero se obtiene el `id_fabricante` asociado a la columna `make`. 

Luego, usando la combinación de `model`, `trim`, `body` e `id_fabricante`, se busca el `id_modelo` correspondiente en el diccionario `map_modelos`. 

Esto es necesario porque el vehículo no se relaciona directamente con el fabricante, sino con un modelo, y el modelo a su vez pertenece a un fabricante.

Finalmente, se construye una lista de tuplas con los datos de cada vehículo y se insertan en la tabla `Vehiculos` mediante `executemany`. 

La instrucción `INSERT OR IGNORE` funciona como una protección adicional frente a posibles duplicados, aunque estos ya fueron controlados previamente al conservar una sola fila por `vin`.


````python
def limpiar_numero(valor): # la usaremos
    if pd.isna(valor):
        return None
    return float(valor)
````

````python
ventas = []

for _, row in df_carga.iterrows():
    id_vendedor = map_vendedores[row["seller"]]
    id_estado = map_estados[row["state"]]
    
    ventas.append((
        row["vin"],
        id_vendedor,
        id_estado,
        limpiar_numero(row["condition"]),
        limpiar_numero(row["odometer"]),
        limpiar_numero(row["mmr"]),
        limpiar_numero(row["sellingprice"]),
        row["saledate"]
    ))

cursor.executemany("""
    INSERT INTO Ventas(
        vin,
        id_vendedor,
        id_estado,
        condition,
        odometer,
        mmr,
        sellingprice,
        saledate
    )
    VALUES (?, ?, ?, ?, ?, ?, ?, ?)
""", ventas)

connection.commit()
````

Finalmente, Ventas, que representa cada registro de venta u observación presente en el CSV original. 

A diferencia de la tabla `Vehiculos`, aquí no se eliminan filas repetidas por `vin`, ya que un mismo vehículo puede aparecer en más de una venta. Por esta razón, se recorre todo `df_carga` fila por fila.

Antes de insertar los datos, se define la función `limpiar_numero`. 

Esta función se usa para transformar valores faltantes de pandas (`NaN`) en `None`, ya que SQLite interpreta `None` como un valor nulo (`NULL`). 

Si el valor no es nulo, se convierte a `float`. Esto permite insertar correctamente columnas numéricas como `condition`, `odometer`, `mmr` y `sellingprice`.

Durante el recorrido de cada fila, se obtiene el `id_vendedor` correspondiente al valor de `seller` usando el diccionario `map_vendedores`, y el `id_estado` correspondiente al valor de `state` usando `map_estados`. 

Esto es necesario porque la tabla `Ventas` no guarda directamente el nombre del vendedor ni el código del estado como texto, sino que guarda llaves foráneas hacia las tablas `Vendedores` y `Estados`.

Luego, para cada fila del CSV, se construye una tupla con los datos necesarios para insertar una venta: el `vin` del vehículo vendido, el identificador del vendedor, el identificador del estado, la condición del vehículo, el odómetro, el valor de referencia `mmr`, el precio de venta y la fecha de venta. Todas estas tuplas se almacenan en la lista `ventas`.


````python
tablas = [
    "Fabricantes",
    "Modelos",
    "Vehiculos",
    "Estados",
    "Vendedores",
    "Ventas"
]

conteos = []

for tabla in tablas:
    cursor.execute(f"SELECT COUNT(*) FROM {tabla}")
    n_registros = cursor.fetchone()[0]
    conteos.append({
        "tabla": tabla,
        "registros_cargados": n_registros
    })

resumen_carga = pd.DataFrame(conteos)
resumen_carga
````

````python
esperados = {
    "Fabricantes": df_carga["make"].nunique(),
    "Modelos": df_carga[["make", "model", "trim", "body"]].drop_duplicates().shape[0],
    "Vehiculos": df_carga["vin"].nunique(),
    "Estados": df_carga["state"].nunique(),
    "Vendedores": df_carga["seller"].nunique(),
    "Ventas": len(df_carga)
}

validacion = []

for tabla, esperado in esperados.items():
    cursor.execute(f"SELECT COUNT(*) FROM {tabla}")
    registros_bd = cursor.fetchone()[0]
    
    validacion.append({
        "tabla": tabla,
        "esperado_desde_dataframe": esperado,
        "registros_cargados_bd": registros_bd,
        "coincide": esperado == registros_bd
    })

resumen_validacion = pd.DataFrame(validacion)
resumen_validacion
````

La tabla de validación compara, para cada entidad, la cantidad esperada de registros calculada desde el DataFrame con la cantidad efectivamente cargada en la base de datos. En todos los casos, ambos valores coinciden.

Esto confirma que la carga fue consistente con el modelo relacional definido.

Las tablas de entidades únicas (`Fabricantes`, `Estados` y `Vendedores`) contienen los valores únicos correspondientes; `Modelos` contiene las combinaciones únicas de fabricante, modelo, versión y carrocería; `Vehiculos` contiene un registro por cada `vin` único; y `Ventas` conserva todas las filas originales del CSV como registros de venta.
