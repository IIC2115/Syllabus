```python
import sqlite3

conn = sqlite3.connect("store.db")
cursor = conn.cursor()

cursor.execute("PRAGMA foreign_keys = ON;")
```

```python
schema = """

-- eliminar tablas si ya existen
-- se eliminan primero las tablas dependientes

DROP TABLE IF EXISTS order_items;
DROP TABLE IF EXISTS stocks;
DROP TABLE IF EXISTS orders;
DROP TABLE IF EXISTS products;
DROP TABLE IF EXISTS staffs;
DROP TABLE IF EXISTS customers;
DROP TABLE IF EXISTS stores;
DROP TABLE IF EXISTS categories;
DROP TABLE IF EXISTS brands;

-- tablas independientes

CREATE TABLE brands (
    brand_id INTEGER PRIMARY KEY,
    brand_name TEXT NOT NULL UNIQUE
);

CREATE TABLE categories (
    category_id INTEGER PRIMARY KEY,
    category_name TEXT NOT NULL UNIQUE
);

CREATE TABLE customers (
    customer_id INTEGER PRIMARY KEY,
    first_name TEXT NOT NULL,
    last_name TEXT NOT NULL,
    phone TEXT,
    email TEXT NOT NULL UNIQUE,
    street TEXT NOT NULL,
    city TEXT NOT NULL,
    state TEXT NOT NULL,
    zip_code TEXT NOT NULL
);

CREATE TABLE stores (
    store_id INTEGER PRIMARY KEY,
    store_name TEXT NOT NULL,
    phone TEXT NOT NULL,
    email TEXT NOT NULL UNIQUE,
    street TEXT NOT NULL,
    city TEXT NOT NULL,
    state TEXT NOT NULL,
    zip_code TEXT NOT NULL
);

-- tabla de empleados
-- depende de stores y también se referencia a sí misma

CREATE TABLE staffs (
    staff_id INTEGER PRIMARY KEY,
    first_name TEXT NOT NULL,
    last_name TEXT NOT NULL,
    email TEXT NOT NULL UNIQUE,
    phone TEXT NOT NULL,
    active INTEGER NOT NULL CHECK (active IN (0, 1)),
    store_id INTEGER NOT NULL,
    manager_id INTEGER,
    FOREIGN KEY (store_id) REFERENCES stores(store_id),
    FOREIGN KEY (manager_id) REFERENCES staffs(staff_id)
);

-- tabla de productos
-- depende de brands y categories

CREATE TABLE products (
    product_id INTEGER PRIMARY KEY,
    product_name TEXT NOT NULL,
    brand_id INTEGER NOT NULL,
    category_id INTEGER NOT NULL,
    model_year INTEGER NOT NULL,
    list_price REAL NOT NULL CHECK (list_price >= 0),
    FOREIGN KEY (brand_id) REFERENCES brands(brand_id),
    FOREIGN KEY (category_id) REFERENCES categories(category_id)
);

-- tabla de órdenes
-- depende de customers, stores y staffs

CREATE TABLE orders (
    order_id INTEGER PRIMARY KEY,
    customer_id INTEGER NOT NULL,
    order_status INTEGER NOT NULL CHECK (order_status IN (1, 2, 3, 4)),
    order_date TEXT NOT NULL,
    required_date TEXT NOT NULL,
    shipped_date TEXT,
    store_id INTEGER NOT NULL,
    staff_id INTEGER NOT NULL,
    FOREIGN KEY (customer_id) REFERENCES customers(customer_id),
    FOREIGN KEY (store_id) REFERENCES stores(store_id),
    FOREIGN KEY (staff_id) REFERENCES staffs(staff_id)
);

-- detalle de órdenes
-- depende de orders y products

CREATE TABLE order_items (
    order_id INTEGER NOT NULL,
    item_id INTEGER NOT NULL,
    product_id INTEGER NOT NULL,
    quantity INTEGER NOT NULL CHECK (quantity > 0),
    list_price REAL NOT NULL CHECK (list_price >= 0),
    discount REAL NOT NULL CHECK (discount >= 0 AND discount <= 1),
    PRIMARY KEY (order_id, item_id),
    FOREIGN KEY (order_id) REFERENCES orders(order_id),
    FOREIGN KEY (product_id) REFERENCES products(product_id)
);

-- stock por tienda y producto
-- depende de stores y products

CREATE TABLE stocks (
    store_id INTEGER NOT NULL,
    product_id INTEGER NOT NULL,
    quantity INTEGER NOT NULL CHECK (quantity >= 0),
    PRIMARY KEY (store_id, product_id),
    FOREIGN KEY (store_id) REFERENCES stores(store_id),
    FOREIGN KEY (product_id) REFERENCES products(product_id)
);

"""

cursor.executescript(schema)
conn.commit()
```

```python
def run_sql(query):
    return pd.read_sql_query(query, conn)

def comparar_resultados(sql_df, py_df, sort_cols=None):

    # compara dos df.
    # si sort_cols se entrega, ordena ambos df antes de comparar.

    sql_comp = sql_df.copy()
    py_comp = py_df.copy()
    
    if sort_cols is not None:
        sql_comp = sql_comp.sort_values(sort_cols).reset_index(drop=True)
        py_comp = py_comp.sort_values(sort_cols).reset_index(drop=True)
    else:
        sql_comp = sql_comp.reset_index(drop=True)
        py_comp = py_comp.reset_index(drop=True)
    
    pd.testing.assert_frame_equal(
        sql_comp,
        py_comp,
        check_dtype=False,
        check_exact=False,
        atol=1e-6,
        rtol=1e-6
    )
    
    print("SQL y Python coinciden.")
```

```python
sql_g = """
WITH ingresos_producto AS (
    SELECT
        p.product_id,
        p.product_name,
        p.category_id,
        c.category_name,
        SUM(oi.quantity * oi.list_price * (1 - oi.discount)) AS ingreso_producto
    FROM products p
    JOIN categories c
        ON p.category_id = c.category_id
    JOIN order_items oi
        ON p.product_id = oi.product_id
    GROUP BY p.product_id, p.product_name, p.category_id, c.category_name
),
promedio_categoria AS (
    SELECT
        category_id,
        AVG(ingreso_producto) AS ingreso_promedio_categoria
    FROM ingresos_producto
    GROUP BY category_id
)
SELECT
    ip.product_id,
    ip.product_name,
    ip.category_id,
    ip.category_name,
    ip.ingreso_producto,
    pc.ingreso_promedio_categoria
FROM ingresos_producto ip
JOIN promedio_categoria pc
    ON ip.category_id = pc.category_id
WHERE ip.ingreso_producto > pc.ingreso_promedio_categoria
ORDER BY ip.category_id ASC, ip.ingreso_producto DESC;
"""

res_sql_g = run_sql(sql_g)
display(res_sql_g)
```

```python
ingresos_producto_py = (
    products
    .merge(categories, on="category_id")
    .merge(order_items, on="product_id")
)

ingresos_producto_py["ingreso_linea"] = (
    ingresos_producto_py["quantity"] 
    * ingresos_producto_py["list_price_y"] 
    * (1 - ingresos_producto_py["discount"])
)

ingresos_producto_py = (
    ingresos_producto_py
    .groupby(
        ["product_id", "product_name", "category_id", "category_name"],
        as_index=False
    )
    .agg(ingreso_producto=("ingreso_linea", "sum"))
)

promedio_categoria_py = (
    ingresos_producto_py
    .groupby("category_id", as_index=False)
    .agg(ingreso_promedio_categoria=("ingreso_producto", "mean"))
)

res_py_g = (
    ingresos_producto_py
    .merge(promedio_categoria_py, on="category_id")
)

res_py_g = (
    res_py_g[res_py_g["ingreso_producto"] > res_py_g["ingreso_promedio_categoria"]]
    .sort_values(["category_id", "ingreso_producto"], ascending=[True, False])
    .reset_index(drop=True)
)

display(res_py_g)

comparar_resultados(res_sql_g, res_py_g)
```

Decimos que una compra cuenta si el producto comprado no registra ventas posteriores a esa fecha.

Es decir, buscamos compras que ocurrieron en la última fecha de venta de cada producto.

```python
sql_h = """
WITH compras AS (
    SELECT
        c.customer_id,
        c.first_name,
        c.last_name,
        o.order_date,
        oi.product_id,
        oi.quantity * oi.list_price * (1 - oi.discount) AS gasto
    FROM customers c
    JOIN orders o
        ON c.customer_id = o.customer_id
    JOIN order_items oi
        ON o.order_id = oi.order_id
),
ultima_venta_producto AS (
    SELECT
        product_id,
        MAX(order_date) AS ultima_fecha_venta
    FROM compras
    GROUP BY product_id
)
SELECT
    compras.customer_id,
    compras.first_name,
    compras.last_name,
    SUM(compras.gasto) AS total_gastado
FROM compras
JOIN ultima_venta_producto uv
    ON compras.product_id = uv.product_id
   AND compras.order_date = uv.ultima_fecha_venta
GROUP BY compras.customer_id, compras.first_name, compras.last_name
ORDER BY total_gastado DESC, compras.customer_id ASC
LIMIT 10;
"""

res_sql_h = run_sql(sql_h)
display(res_sql_h)
```

```python
compras_h = (
    customers
    .merge(orders, on="customer_id")
    .merge(order_items, on="order_id")
)

compras_h["gasto"] = (
    compras_h["quantity"] * compras_h["list_price"] * (1 - compras_h["discount"])
)

ultima_venta_producto_py = (
    compras_h
    .groupby("product_id", as_index=False)
    .agg(ultima_fecha_venta=("order_date", "max"))
)

compras_h = compras_h.merge(
    ultima_venta_producto_py,
    on="product_id"
)

compras_h_filtrado = compras_h[
    compras_h["order_date"] == compras_h["ultima_fecha_venta"]
]

res_py_h = (
    compras_h_filtrado
    .groupby(["customer_id", "first_name", "last_name"], as_index=False)
    .agg(total_gastado=("gasto", "sum"))
    .sort_values(["total_gastado", "customer_id"], ascending=[False, True])
    .head(10)
    .reset_index(drop=True)
)

display(res_py_h)

comparar_resultados(res_sql_h, res_py_h)
```

i) comparamos stock total disponible > unidades totales vendidas

El stock se suma desde stocks, y las ventas acumuladas desde order_items.

```python
sql_i = """
WITH ventas_producto AS (
    SELECT
        product_id,
        SUM(quantity) AS unidades_vendidas
    FROM order_items
    GROUP BY product_id
),
stock_producto AS (
    SELECT
        product_id,
        SUM(quantity) AS unidades_stock
    FROM stocks
    GROUP BY product_id
)
SELECT
    p.product_id,
    p.product_name,
    COALESCE(sp.unidades_stock, 0) AS unidades_stock,
    COALESCE(vp.unidades_vendidas, 0) AS unidades_vendidas
FROM products p
LEFT JOIN stock_producto sp
    ON p.product_id = sp.product_id
LEFT JOIN ventas_producto vp
    ON p.product_id = vp.product_id
WHERE COALESCE(sp.unidades_stock, 0) > COALESCE(vp.unidades_vendidas, 0)
ORDER BY unidades_stock DESC, p.product_id ASC;
"""

res_sql_i = run_sql(sql_i)
display(res_sql_i)
```

```python
ventas_producto_py = (
    order_items
    .groupby("product_id", as_index=False)
    .agg(unidades_vendidas=("quantity", "sum"))
)

stock_producto_py = (
    stocks
    .groupby("product_id", as_index=False)
    .agg(unidades_stock=("quantity", "sum"))
)

res_py_i = (
    products[["product_id", "product_name"]]
    .merge(stock_producto_py, on="product_id", how="left")
    .merge(ventas_producto_py, on="product_id", how="left")
)

res_py_i["unidades_stock"] = res_py_i["unidades_stock"].fillna(0)
res_py_i["unidades_vendidas"] = res_py_i["unidades_vendidas"].fillna(0)

res_py_i = (
    res_py_i[res_py_i["unidades_stock"] > res_py_i["unidades_vendidas"]]
    .sort_values(["unidades_stock", "product_id"], ascending=[False, True])
    .reset_index(drop=True)
)

display(res_py_i)

comparar_resultados(res_sql_i, res_py_i)
```

j) Primero calculamos ventas por tienda y mes. Luego usamos el mes anterior para calcular variación % = (total_mes_actual - total_mes_anterior) / total_mes_anterior * 100

En SQL usamos LAG(). Vean ese comando!

```python
sql_j = """
WITH ventas_mensuales AS (
    SELECT
        s.store_id,
        s.store_name,
        strftime('%Y-%m', o.order_date) AS mes,
        SUM(oi.quantity * oi.list_price * (1 - oi.discount)) AS total_vendido
    FROM stores s
    JOIN orders o
        ON s.store_id = o.store_id
    JOIN order_items oi
        ON o.order_id = oi.order_id
    GROUP BY s.store_id, s.store_name, mes
),
ventas_con_lag AS (
    SELECT
        store_id,
        store_name,
        mes,
        total_vendido,
        LAG(total_vendido) OVER (
            PARTITION BY store_id
            ORDER BY mes
        ) AS total_mes_anterior
    FROM ventas_mensuales
)
SELECT
    store_id,
    store_name,
    mes,
    total_vendido,
    total_mes_anterior,
    CASE
        WHEN total_mes_anterior IS NULL OR total_mes_anterior = 0 THEN NULL
        ELSE 100.0 * (total_vendido - total_mes_anterior) / total_mes_anterior
    END AS variacion_porcentual
FROM ventas_con_lag
ORDER BY store_id ASC, mes ASC;
"""

res_sql_j = run_sql(sql_j)
display(res_sql_j)
```

```python
orders_j = orders.copy()
orders_j["mes"] = orders_j["order_date"].dt.to_period("M").astype(str)

ventas_j = (
    stores
    .merge(orders_j, on="store_id")
    .merge(order_items, on="order_id")
)

ventas_j["total_linea"] = (
    ventas_j["quantity"] * ventas_j["list_price"] * (1 - ventas_j["discount"])
)

res_py_j = (
    ventas_j
    .groupby(["store_id", "store_name", "mes"], as_index=False)
    .agg(total_vendido=("total_linea", "sum"))
    .sort_values(["store_id", "mes"])
    .reset_index(drop=True)
)

res_py_j["total_mes_anterior"] = (
    res_py_j
    .groupby("store_id")["total_vendido"]
    .shift(1)
)

res_py_j["variacion_porcentual"] = (
    100 * 
    (res_py_j["total_vendido"] - res_py_j["total_mes_anterior"]) /
    res_py_j["total_mes_anterior"]
)

res_py_j.loc[
    res_py_j["total_mes_anterior"].isna() | (res_py_j["total_mes_anterior"] == 0),
    "variacion_porcentual"
] = np.nan

display(res_py_j)

comparar_resultados(res_sql_j, res_py_j)
```

```python
sql_l = """
WITH categorias_cliente AS (
    SELECT
        o.customer_id,
        COUNT(DISTINCT p.category_id) AS categorias_compradas
    FROM orders o
    JOIN order_items oi
        ON o.order_id = oi.order_id
    JOIN products p
        ON oi.product_id = p.product_id
    GROUP BY o.customer_id
),
total_categorias AS (
    SELECT
        COUNT(*) AS total
    FROM categories
)
SELECT
    c.customer_id,
    c.first_name,
    c.last_name,
    cc.categorias_compradas
FROM customers c
JOIN categorias_cliente cc
    ON c.customer_id = cc.customer_id
CROSS JOIN total_categorias tc
WHERE cc.categorias_compradas = tc.total
ORDER BY c.customer_id ASC;
"""

res_sql_l = run_sql(sql_l)
display(res_sql_l)
```

```python
sql_k = """
WITH precio_mensual AS (
    SELECT
        oi.product_id,
        p.product_name,
        strftime('%Y-%m', o.order_date) AS mes,
        date(strftime('%Y-%m', o.order_date) || '-01') AS mes_date,
        SUM(oi.quantity * oi.list_price * (1 - oi.discount)) * 1.0 / SUM(oi.quantity) AS precio_promedio
    FROM order_items oi
    JOIN orders o
        ON oi.order_id = o.order_id
    JOIN products p
        ON oi.product_id = p.product_id
    GROUP BY oi.product_id, p.product_name, mes, mes_date
),
comparacion AS (
    SELECT
        product_id,
        product_name,
        mes,
        mes_date,
        precio_promedio,
        LAG(precio_promedio) OVER (
            PARTITION BY product_id
            ORDER BY mes_date
        ) AS precio_anterior,
        LAG(mes_date) OVER (
            PARTITION BY product_id
            ORDER BY mes_date
        ) AS mes_anterior
    FROM precio_mensual
),
marcas AS (
    SELECT
        *,
        CASE
            WHEN precio_promedio < precio_anterior
             AND mes_date = date(mes_anterior, '+1 month')
            THEN 1
            ELSE 0
        END AS baja_consecutiva
    FROM comparacion
),
grupos AS (
    SELECT
        *,
        SUM(
            CASE 
                WHEN baja_consecutiva = 0 THEN 1 
                ELSE 0 
            END
        ) OVER (
            PARTITION BY product_id
            ORDER BY mes_date
            ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW
        ) AS grupo_racha
    FROM marcas
),
rachas AS (
    SELECT
        product_id,
        product_name,
        grupo_racha,
        COUNT(*) AS meses_con_baja_consecutiva
    FROM grupos
    WHERE baja_consecutiva = 1
    GROUP BY product_id, product_name, grupo_racha
),
max_rachas AS (
    SELECT
        product_id,
        product_name,
        MAX(meses_con_baja_consecutiva) AS max_bajas_consecutivas
    FROM rachas
    GROUP BY product_id, product_name
)
SELECT
    product_id,
    product_name,
    max_bajas_consecutivas
FROM max_rachas
WHERE max_bajas_consecutivas >= 3
ORDER BY max_bajas_consecutivas DESC, product_id ASC;
"""

res_sql_k = run_sql(sql_k)
display(res_sql_k)
```

```python
ventas_k = (
    order_items
    .merge(orders[["order_id", "order_date"]], on="order_id")
    .merge(products[["product_id", "product_name"]], on="product_id")
)

ventas_k["mes"] = ventas_k["order_date"].dt.to_period("M").astype(str)
ventas_k["mes_date"] = ventas_k["order_date"].dt.to_period("M").dt.to_timestamp()

ventas_k["ingreso_linea"] = (
    ventas_k["quantity"] * ventas_k["list_price"] * (1 - ventas_k["discount"])
)

precio_mensual_py = (
    ventas_k
    .groupby(["product_id", "product_name", "mes", "mes_date"], as_index=False)
    .agg(
        ingreso_total=("ingreso_linea", "sum"),
        unidades=("quantity", "sum")
    )
)

precio_mensual_py["precio_promedio"] = (
    precio_mensual_py["ingreso_total"] / precio_mensual_py["unidades"]
)

precio_mensual_py = precio_mensual_py.sort_values(["product_id", "mes_date"])

precio_mensual_py["precio_anterior"] = (
    precio_mensual_py
    .groupby("product_id")["precio_promedio"]
    .shift(1)
)

precio_mensual_py["mes_anterior"] = (
    precio_mensual_py
    .groupby("product_id")["mes_date"]
    .shift(1)
)

precio_mensual_py["baja_consecutiva"] = (
    (precio_mensual_py["precio_promedio"] < precio_mensual_py["precio_anterior"]) &
    (precio_mensual_py["mes_date"] == precio_mensual_py["mes_anterior"] + pd.DateOffset(months=1))
).astype(int)

precio_mensual_py["corte_racha"] = (precio_mensual_py["baja_consecutiva"] == 0).astype(int)

precio_mensual_py["grupo_racha"] = (
    precio_mensual_py
    .groupby("product_id")["corte_racha"]
    .cumsum()
)

rachas_py = (
    precio_mensual_py[precio_mensual_py["baja_consecutiva"] == 1]
    .groupby(["product_id", "product_name", "grupo_racha"], as_index=False)
    .size()
    .rename(columns={"size": "meses_con_baja_consecutiva"})
)

res_py_k = (
    rachas_py
    .groupby(["product_id", "product_name"], as_index=False)
    .agg(max_bajas_consecutivas=("meses_con_baja_consecutiva", "max"))
)

res_py_k = (
    res_py_k[res_py_k["max_bajas_consecutivas"] >= 3]
    .sort_values(["max_bajas_consecutivas", "product_id"], ascending=[False, True])
    .reset_index(drop=True)
)

display(res_py_k)

comparar_resultados(res_sql_k, res_py_k)
```

```python
sql_m = """
WITH dias_venta AS (
    SELECT DISTINCT
        store_id,
        date(order_date) AS dia
    FROM orders
),
numerados AS (
    SELECT
        store_id,
        dia,
        ROW_NUMBER() OVER (
            PARTITION BY store_id
            ORDER BY dia
        ) AS rn
    FROM dias_venta
),
grupos AS (
    SELECT
        store_id,
        dia,
        date(dia, '-' || rn || ' days') AS grupo
    FROM numerados
),
rachas AS (
    SELECT
        store_id,
        grupo,
        COUNT(*) AS dias_consecutivos
    FROM grupos
    GROUP BY store_id, grupo
),
max_rachas AS (
    SELECT
        store_id,
        MAX(dias_consecutivos) AS racha_maxima_dias
    FROM rachas
    GROUP BY store_id
)
SELECT
    s.store_id,
    s.store_name,
    mr.racha_maxima_dias
FROM max_rachas mr
JOIN stores s
    ON mr.store_id = s.store_id
ORDER BY mr.racha_maxima_dias DESC, s.store_id ASC;
"""

res_sql_m = run_sql(sql_m)
display(res_sql_m)
```

```python
dias_venta_py = (
    orders[["store_id", "order_date"]]
    .copy()
)

dias_venta_py["dia"] = dias_venta_py["order_date"].dt.normalize()

dias_venta_py = (
    dias_venta_py[["store_id", "dia"]]
    .drop_duplicates()
    .sort_values(["store_id", "dia"])
    .reset_index(drop=True)
)

dias_venta_py["rn"] = (
    dias_venta_py
    .groupby("store_id")
    .cumcount() + 1
)

dias_venta_py["grupo"] = (
    dias_venta_py["dia"] - pd.to_timedelta(dias_venta_py["rn"], unit="D")
)

rachas_py = (
    dias_venta_py
    .groupby(["store_id", "grupo"], as_index=False)
    .agg(dias_consecutivos=("dia", "count"))
)

max_rachas_py = (
    rachas_py
    .groupby("store_id", as_index=False)
    .agg(racha_maxima_dias=("dias_consecutivos", "max"))
)

res_py_m = (
    max_rachas_py
    .merge(stores[["store_id", "store_name"]], on="store_id")
    [["store_id", "store_name", "racha_maxima_dias"]]
    .sort_values(["racha_maxima_dias", "store_id"], ascending=[False, True])
    .reset_index(drop=True)
)

display(res_py_m)

comparar_resultados(res_sql_m, res_py_m)
```

```python
sql_n = """
WITH ingresos_mensuales AS (
    SELECT
        strftime('%Y-%m', o.order_date) AS mes,
        oi.product_id,
        p.product_name,
        SUM(oi.quantity * oi.list_price * (1 - oi.discount)) AS ingreso_mensual
    FROM order_items oi
    JOIN orders o
        ON oi.order_id = o.order_id
    JOIN products p
        ON oi.product_id = p.product_id
    GROUP BY mes, oi.product_id, p.product_name
),
ranking_mensual AS (
    SELECT
        mes,
        product_id,
        product_name,
        ingreso_mensual,
        RANK() OVER (
            PARTITION BY mes
            ORDER BY ingreso_mensual DESC
        ) AS ranking_mes
    FROM ingresos_mensuales
),
top1 AS (
    SELECT
        *
    FROM ranking_mensual
    WHERE ranking_mes = 1
)
SELECT
    product_id,
    product_name,
    COUNT(*) AS meses_top1
FROM top1
GROUP BY product_id, product_name
HAVING COUNT(*) >= 6
ORDER BY meses_top1 DESC, product_id ASC;
"""

res_sql_n = run_sql(sql_n)
display(res_sql_n)
```

```python
ventas_n = (
    order_items
    .merge(orders[["order_id", "order_date"]], on="order_id")
    .merge(products[["product_id", "product_name"]], on="product_id")
)

ventas_n["mes"] = ventas_n["order_date"].dt.to_period("M").astype(str)

ventas_n["ingreso_linea"] = (
    ventas_n["quantity"] * ventas_n["list_price"] * (1 - ventas_n["discount"])
)

ingresos_mensuales_py = (
    ventas_n
    .groupby(["mes", "product_id", "product_name"], as_index=False)
    .agg(ingreso_mensual=("ingreso_linea", "sum"))
)

ingresos_mensuales_py["ranking_mes"] = (
    ingresos_mensuales_py
    .groupby("mes")["ingreso_mensual"]
    .rank(method="min", ascending=False)
)

top1_py = ingresos_mensuales_py[
    ingresos_mensuales_py["ranking_mes"] == 1
]

res_py_n = (
    top1_py
    .groupby(["product_id", "product_name"], as_index=False)
    .agg(meses_top1=("mes", "count"))
)

res_py_n = (
    res_py_n[res_py_n["meses_top1"] >= 6]
    .sort_values(["meses_top1", "product_id"], ascending=[False, True])
    .reset_index(drop=True)
)

display(res_py_n)

comparar_resultados(res_sql_n, res_py_n)
```

```python
sql_o = """
WITH ingresos_producto AS (
    SELECT
        p.product_id,
        p.product_name,
        SUM(oi.quantity * oi.list_price * (1 - oi.discount)) AS ingreso_producto
    FROM products p
    JOIN order_items oi
        ON p.product_id = oi.product_id
    GROUP BY p.product_id, p.product_name
),
acumulado AS (
    SELECT
        product_id,
        product_name,
        ingreso_producto,
        SUM(ingreso_producto) OVER (
            ORDER BY ingreso_producto DESC, product_id ASC
            ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW
        ) AS ingreso_acumulado,
        SUM(ingreso_producto) OVER () AS ingreso_total
    FROM ingresos_producto
),
resultado AS (
    SELECT
        product_id,
        product_name,
        ingreso_producto,
        ingreso_acumulado,
        ingreso_total,
        ingreso_acumulado - ingreso_producto AS ingreso_acumulado_anterior,
        ingreso_acumulado * 1.0 / ingreso_total AS proporcion_acumulada
    FROM acumulado
)
SELECT
    product_id,
    product_name,
    ingreso_producto,
    ingreso_acumulado,
    ingreso_total,
    proporcion_acumulada
FROM resultado
WHERE ingreso_acumulado_anterior < 0.8 * ingreso_total
ORDER BY ingreso_producto DESC, product_id ASC;
"""

res_sql_o = run_sql(sql_o)
display(res_sql_o)
```

```python
ventas_o = (
    products[["product_id", "product_name"]]
    .merge(order_items, on="product_id")
)

ventas_o["ingreso_linea"] = (
    ventas_o["quantity"] * ventas_o["list_price"] * (1 - ventas_o["discount"])
)

ingresos_producto_py = (
    ventas_o
    .groupby(["product_id", "product_name"], as_index=False)
    .agg(ingreso_producto=("ingreso_linea", "sum"))
)

ingresos_producto_py = (
    ingresos_producto_py
    .sort_values(["ingreso_producto", "product_id"], ascending=[False, True])
    .reset_index(drop=True)
)

ingresos_producto_py["ingreso_acumulado"] = (
    ingresos_producto_py["ingreso_producto"].cumsum()
)

ingresos_producto_py["ingreso_total"] = (
    ingresos_producto_py["ingreso_producto"].sum()
)

ingresos_producto_py["ingreso_acumulado_anterior"] = (
    ingresos_producto_py["ingreso_acumulado"] - ingresos_producto_py["ingreso_producto"]
)

ingresos_producto_py["proporcion_acumulada"] = (
    ingresos_producto_py["ingreso_acumulado"] / ingresos_producto_py["ingreso_total"]
)

res_py_o = (
    ingresos_producto_py[
        ingresos_producto_py["ingreso_acumulado_anterior"] < 
        0.8 * ingresos_producto_py["ingreso_total"]
    ]
    [[
        "product_id",
        "product_name",
        "ingreso_producto",
        "ingreso_acumulado",
        "ingreso_total",
        "proporcion_acumulada"
    ]]
    .reset_index(drop=True)
)

display(res_py_o)

comparar_resultados(res_sql_o, res_py_o)
```
