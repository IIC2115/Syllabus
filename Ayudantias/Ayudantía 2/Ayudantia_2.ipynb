## Entidades de la torre

```python
from random import randint 

class Torre:
    def __init__(self, nombre:str,espacios:dict,estacionamiento:list):
        self.nombre = nombre
        self.espacios = espacios #la idea es hacer un diccionario donde las llaves sean los tipos de espacios,comerical, residencial, etc, 
                                 #y los valores sean listas de objetos de la clase Espacio
        self.altura = self.calculo_altura()
        self.pisos = self.multiplos_de_cinco(randint(40, 100))
        self.espacios_por_piso = self.calcular_espacios_por_piso()
        self.estacionamiento = estacionamiento #Lista de objetos de la clase Vehiculo
        self.capacidad_estacionamiento = self.multiplos_de_cinco(randint(200, 300))
        self.ingresos_estacionamiento = self.calcular_ingresos_estacionamiento()
        self.gastos_estacionamiento = self.calcular_gastos_estacionamiento()
        self.utilidades = self.ingresos_estacionamiento - self.gastos_estacionamiento


    def calculo_altura(self):
        altura = 0
        for lista in list(self.espacios.values()):
            for espacio in lista:
                altura += espacio.altura
        return altura

    def calcular_espacios_por_piso(self):
        total_espacios = 0
        for lista in list(self.espacios.values()):
            total_espacios += len(lista)
        espacios_por_piso = total_espacios // self.pisos
        return espacios_por_piso
       
        
    def multiplos_de_cinco(self,numero):
        if numero%5 != 0: # Esto nos entrega el resto de una division entera, si da 0, numero es multiplo de cinco. 
                          # Si no es 0, lo que entregá es el resto de la divisón entera, es decir, lo que falta para que sea un múltiplo de cinco.
                          # Por ejemplo, si numero es 97, y 97 / 5 = 19.4, dado que son enteros, es 19. Lo que falta para que dicho resultado calcé con el valor original, 
                          # es decir, 97, es 2, porque: (19 * 5) + 2 = 97. Es por eso que es 0 al hacer a%b cuando a y b son multiplos, 
                          # porque no falta ningun muro por sumar, no hay ningun restante 
            numero += 5 - (numero % 5) #Auí por último le sumamos 5 menos el resto para, simplemente, transformar el numero en el numero que sea multiplo de cinco
            return numero
        else:
            return numero

    def calcular_ingresos_estacionamiento(self):
        ingresos = 0
        for vehiculo in self.estacionamiento:
            ingresos += vehiculo.ingreso
        return ingresos

    def calcular_gastos_estacionamiento(self):
        gastos = 0
        for vehiculo in self.estacionamiento:
            gastos += vehiculo.gasto
        return gastos


    def ocupacion_estacionamiento(self):
        return len(self.estacionamiento) / self.capacidad_estacionamiento if self.capacidad_estacionamiento > 0 else 0

    def __str__(self):
        return f"Torre {self.nombre}: {self.pisos} pisos. \n Espacios por piso: {self.espacios_por_piso}, Capacidad de estacionamiento: {self.capacidad_estacionamiento}. \n Ocupación estacionamiento: {self.ocupacion_estacionamiento()}"
```

## Entidades de los espacios de la torre

```python
from abc import ABC, abstractmethod 

from random import randint # No es necesario volver a escribir esta línea, sin embargo, lo escribo de todas formas

class Espacio(ABC):
    @abstractmethod
    def __init__(self, nombre:str,categoria:str,tipo:str):
        self.nombre = nombre
        self.categoria = categoria #Posibles categorias: Residencial, Oficina, Comercial. 
                                    #Para tener la futura estructura de datos más ordenada
        self.tipo = tipo
        self.espacio = self.calcular_espacio()
        self.ingresos = self.calcular_ingresos()
        self.gastos = self.calcular_gastos()
        self.utilidades = self.ingresos - self.gastos 
        self.altura = 0 #La altura se define en las clases hijas, dependiendo de la categoría del espacio.
        

    @abstractmethod
    def calcular_espacio(self):
        pass

    @abstractmethod
    def calcular_ingresos(self):
        pass
    
    @abstractmethod
    def calcular_gastos(self):
        pass


class Residencial(Espacio):
    def __init__(self, nombre:str, categoria:str,tipo:str):
        super().__init__(nombre,categoria,tipo)
        if self.categoria.lower() != "residencial": #Para evitar errores de que se asigne una categoría invalida
            raise ValueError("La categoria del espacio no es valida")
        self.altura = 5
        
    def calcular_espacio(self):
        if self.tipo.lower() == "departamento pequeño":
            return 2

        elif self.tipo.lower() == "departamento mediano":
            return 4

        elif self.tipo.lower() == "departamento grande":
            return randint(6, 10)

        elif self.tipo.lower() == "departamento hotelero": #Un departamento que solo se arriende por una noche o a lo más cuatro
            return 2
        
        else:
            raise ValueError("El tipo de departamento no es valido")
            


    def calcular_ingresos(self):
        if self.tipo.lower() == "departamento pequeño":
            return 5000

        elif self.tipo.lower() == "departamento mediano":
            return 7000

        elif self.tipo.lower() == "departamento grande":
            return 10000

        elif self.tipo.lower() == "departamento hotelero": #Un departamento que solo se arriende por una noche o a lo más cuatro
            return 12000
        
        else:
            raise ValueError("El tipo de departamento no es valido")

    def calcular_gastos(self):
        if self.tipo.lower() == "departamento pequeño":
            return 2000
        elif self.tipo.lower() == "departamento mediano":
            return 3000
        elif self.tipo.lower() == "departamento grande":
            return 4000
        elif self.tipo.lower() == "departamento hotelero":
            return 5000
        else:
            raise ValueError("El tipo de departamento no es valido")
    

class Oficina(Espacio):
    def __init__(self, nombre:str,categoria:str,tipo:str):
        super().__init__(nombre,categoria,tipo)
        if self.categoria.lower() != "oficina": #Para evitar errores de que se asigne una categoría invalida
            raise ValueError("La categoria del espacio no es valida")
        self.altura = 10
    
    
    
    def calcular_espacio(self):
        if self.tipo.lower() == "oficina":
            return 1

        elif self.tipo.lower() == "oficina de reuniones":
            return 12

        elif self.tipo.lower() == "oficina de coworking":
            return randint(12,20)
        
        else:
            raise ValueError("El tipo oficina ingresado no es valido")

    
    
    def calcular_ingresos(self):
        if self.tipo.lower() == "oficina":
            return 10000

        elif self.tipo.lower() == "oficina de reuniones":
            return 20000

        elif self.tipo.lower() == "oficina de coworking":
            return 30000
        
        else:
            raise ValueError("El tipo oficina ingresado no es valido")
    
    
    def calcular_gastos(self): #Para el caso de los gastos, asumiremos que solo gasta por limpieza, agua y electricidad
        if self.tipo.lower() == "oficina":
            return 5000
        elif self.tipo.lower() == "oficina de reuniones":
            return 10000
        elif self.tipo.lower() == "oficina de coworking":
            return 15000
        else:
            raise ValueError("El tipo oficina no es valido")


class Comercial(Espacio):
    def __init__(self, nombre:str,categoria:str,tipo:str,comunitario:bool):
        self.comunitario = comunitario #Agregamos si se hace servicio comunitario o no. 
        super().__init__(nombre,categoria,tipo)
        if self.categoria.lower() != "comercial": #Para evitar errores de que se asigne una categoría invalida
            raise ValueError("La categoria del espacio no es valida")
        self.altura = 15
        
    
    def calcular_espacio(self):
        if self.tipo.lower() == "tienda":
            return randint(5, 10)
        elif self.tipo.lower() == "restaurante":
            return randint(20, 50)
        elif self.tipo.lower() == "supermercado":
            return randint(30, 70)
        elif self.tipo.lower() == "farmacia":
            return randint(5, 10)
        elif self.tipo.lower() == "libreria":
            return randint(3, 6)
        else:
            raise ValueError("El tipo de comercio ingresado no es valido")
    


    def calcular_ingresos(self):
        if self.tipo.lower() == "tienda": 
            return 1500
        elif self.tipo.lower() == "restaurante":
            return 30000

        elif self.tipo.lower() == "supermercado":
            return 7000

        elif self.tipo.lower() == "farmacia":
            return 2000

        elif self.tipo.lower() == "libreria":
            return 2000
        
        else:
            raise ValueError("El tipo de comercio no es valido")
        


    def calcular_gastos(self):
        if self.tipo.lower() == "tienda":
            if self.comunitario:
                return 1200
            
            else:
                return 1000
        elif self.tipo.lower() == "restaurante":
            if self.comunitario:
                return 18000
            else:
                return 20000

        elif self.tipo.lower() == "supermercado":
            if self.comunitario:
                return 6000
            else:
                return 6000

        elif self.tipo.lower() == "farmacia":
            if self.comunitario:
                return 1500
            else:
                return 2000
        elif self.tipo.lower() == "libreria":
            if self.comunitario:
                return 1500
            else:
                return 2000
        else:
            raise ValueError("El tipo de comercio no es valido")
```

## Entidades de las personas

```python
class Persona(ABC):
    @abstractmethod
    def __init__(self, nombre:str, genero:str):
        self.nombre = nombre
        self.genero = genero

class Residente(Persona):
    def __init__(self, nombre:str, genero:str):
        super().__init__(nombre, genero)
        self.espacio = 1
        self.tipo = "residente"

class Trabajador(Persona):
    def __init__(self, nombre:str, genero:str):
        super().__init__(nombre, genero)
        self.espacio = 3
        self.tipo = "trabajador"
        

class Visitante(Persona):
    def __init__(self, nombre:str, genero:str):
        super().__init__(nombre, genero)
        self.espacio = 2
        self.tipo = "visitante"
```

```python
from abc import ABC, abstractmethod

class Vehiculo(ABC):
    def __init__(self, nombre, marca:str, modelo:str, anio:int, propietario):
        self.nombre = nombre
        self.marca = marca
        self.modelo = modelo
        self.anio = anio #Para evitar problemas en vez de usar la "ñ"
        self.propietario = propietario #Aca la idea será que el propietario sea un objeto de alguna de las clase sherededadas de "Persona" 
        self.ingreso = self.calcular_ingreso() #Nos referimos al ingreso por usar el estacionamiento. 
        self.gasto = self.calcular_gasto() #Nos referimos a los gastos que se tiene por darle un espacio para estacionar a este auto

    @abstractmethod
    def calcular_ingreso(self):
        pass

    @abstractmethod
    def calcular_gasto(self):
        pass
    

    

class Auto(Vehiculo):

    def calcular_ingreso(self):
        if self.propietario.tipo.lower() == "residente":
            return 1000

        elif self.propietario.tipo.lower() == "trabajador":
            return 500

        elif self.propietario.tipo.lower() == "visitante":
            return 3000
        
        else:
            raise ValueError("Tipo de propietario no valido")

    def calcular_gasto(self):
        if self.propietario.tipo.lower() == "residente":
            return 400

        elif self.propietario.tipo.lower() == "trabajador":
            return 250

        elif self.propietario.tipo.lower() == "visitante":
            return 100
        
        else:
            raise ValueError("Tipo de propietario no valido")
        

class Camioneta(Vehiculo):

    def calcular_ingreso(self):
        if self.propietario.tipo.lower() == "residente":
            return 1200

        elif self.propietario.tipo.lower() == "trabajador":
            return 550

        elif self.propietario.tipo.lower() == "visitante":
            return 3500
        
        else:
            raise ValueError("Tipo de propietario no valido")

    def calcular_gasto(self):
        if self.propietario.tipo.lower() == "residente":
            return 420

        elif self.propietario.tipo.lower() == "trabajador":
            return 270

        elif self.propietario.tipo.lower() == "visitante":
            return 120
        
        else:
            raise ValueError("Tipo de propietario no valido")
        
    

class Moto(Vehiculo):

    def calcular_ingreso(self):
        if self.propietario.tipo.lower() == "residente":
            return 700

        elif self.propietario.tipo.lower() == "trabajador":
            return 350

        elif self.propietario.tipo.lower() == "visitante":
            return 2500
        
        else:
            raise ValueError("Tipo de propietario no valido")

    def calcular_gasto(self):
        if self.propietario.tipo.lower() == "residente":
            return 150

        elif self.propietario.tipo.lower() == "trabajador":
            return 100

        elif self.propietario.tipo.lower() == "visitante":
            return 50
        
        else:
            raise ValueError("Tipo de propietario no valido")
        
        

class Camion(Vehiculo): #Hacemos el supuesto obviamente de que son solo camiones pequeños

    def calcular_ingreso(self):
        if self.propietario.tipo.lower() == "residente":
            return 2000

        elif self.propietario.tipo.lower() == "trabajador":
            return 1200

        elif self.propietario.tipo.lower() == "visitante":
            return 7000
        
        else:
            raise ValueError("Tipo de propietario no valido")

    def calcular_gasto(self):
        if self.propietario.tipo.lower() == "residente":
            return 800

        elif self.propietario.tipo.lower() == "trabajador":
            return 400

        elif self.propietario.tipo.lower() == "visitante":
            return 300
        
        else:
            raise ValueError("Tipo de propietario no valido")
```

```python
import unittest
from unittest.mock import patch


# Helpers: Usaremos helpers, es decir, clases de prueba unicamente para probar ciertas clases dentro de nuestros test, para ver si funcionan, en este caso, la clase Persona y
#Vehiculo, que por lo demas eran abstractas. 

class PruebaPersona(Persona):
    def __init__(self, nombre="X", genero="M", tipo="otro"):
        super().__init__(nombre, genero)
        self.tipo = tipo
        self.espacio = 0

class PruebaVehiculo(Vehiculo):
    # No se usa en producción, solo para probar sumas en Torre
    def __init__(self, ingreso, gasto):
       
        self._ingreso = ingreso
        self._gasto = gasto
        super().__init__("Persona de Prueba", "marca", "modelo", 2000, PruebaPersona())

    def calcular_ingreso(self):
        return self._ingreso

    def calcular_gasto(self):
        return self._gasto



# Tests para Espacios

class TestResidencial(unittest.TestCase):
    def test_residencial_constantes_y_altura(self):
        r1 = Residencial("R1", "Residencial", "departamento pequeño")
        self.assertEqual(r1.altura, 5) #assertEqual(a, b) -> Verifica que a == b. Si no son iguales, el test falla.
        self.assertEqual(r1.calcular_espacio(), 2)
        self.assertEqual(r1.calcular_ingresos(), 5000)
        self.assertEqual(r1.calcular_gastos(), 2000)
        self.assertEqual(r1.utilidades, 5000 - 2000)

        r2 = Residencial("R2", "Residencial", "departamento mediano")
        self.assertEqual(r2.calcular_espacio(), 4)
        self.assertEqual(r2.calcular_ingresos(), 7000)
        self.assertEqual(r2.calcular_gastos(), 3000)

        r3 = Residencial("R3", "Residencial", "departamento hotelero")
        self.assertEqual(r3.calcular_espacio(), 2)
        self.assertEqual(r3.calcular_ingresos(), 12000)
        self.assertEqual(r3.calcular_gastos(), 5000)

    def test_residencial_tipo_invalido(self):
        with self.assertRaises(ValueError): #assertRaises(ErrorEsperado, funcion, *args, **kwargs) -> Verifica que una función lance la excepción indicada al ejecutarse.
            Residencial("Rbad", "Residencial", "penthouse imposible")

    def test_residencial_categoria_invalida(self):
        with self.assertRaises(ValueError):
            Residencial("Rbad", "Oficina", "departamento pequeño")


class TestOficina(unittest.TestCase):
    def test_oficina_constantes_y_altura(self):
        o1 = Oficina("O1", "Oficina", "oficina")
        self.assertEqual(o1.altura, 10)
        self.assertEqual(o1.calcular_espacio(), 1)
        self.assertEqual(o1.calcular_ingresos(), 10000)
        self.assertEqual(o1.calcular_gastos(), 5000)

        o2 = Oficina("O2", "Oficina", "oficina de reuniones")
        self.assertEqual(o2.calcular_espacio(), 12)
        self.assertEqual(o2.calcular_ingresos(), 20000)
        self.assertEqual(o2.calcular_gastos(), 10000)

    def test_oficina_tipo_invalido(self):
        with self.assertRaises(ValueError):
            Oficina("Bad", "Oficina", "tipo desconocido")

    def test_oficina_categoria_invalida(self):
        with self.assertRaises(ValueError):
            Oficina("Bad", "Residencial", "oficina")


class TestComercial(unittest.TestCase):
    # Parcheamos __main__.randint, que es reemplazar dicha función (tambien se puede con atributos) de manera temporal durante un test. Para eso usamos @patch, que es una
    # herramienta de la librería de mock, de unittest "unittest.mock", para evitar aleatoriedad y hacer los tests deterministas y reproducibles, vale decir, que siempre 
    # da el mismo resultado si no cambias el código, y además que cualquiera, en cualquier computador, pueda ejecutar el test y obtener el mismo resultado.
    
    
    # En este caso, si hubiesemos importado randint de otra manera, tendríamos que cambiar la ruta en el patch, vale decir, si hubiesemos hecho:
    # import random -> y luego en el codigo usar randint de la forma: random.randint(), el parcheo hubieso incluido a random, vale decir: 
    # @patch("random.randint", side_effect=[41, 203]). Por otro lado, side_effect nos permite definir el comportamiento de la función parcheada. En este caso
    # que es un randint(), el primero valor de la lista será el primer numero que entregará la función  cuando se use, y el segundo el siguiente. Por lo tanto, 
    # si queremos usar el randint(1,100) para que me entregue un valor al azar entre 1 y 100, por el parcheo entregará primero 41, y después 203. 
    
    #Por ultimo, este parcheo que les expliqué, hace efecto en todas las llamadas a randint dentro del código que se ejecute en ese test.
    #Por ejemplo, en esta función "test_torre_atributos_basicos_y_calculos(self, _rnd)" hará efecto al instanciar Torre, por el hecho de que es una clase
    #donde se utiliza randint, y así sucesivamente con las demas instancias. 
    
    @patch("__main__.randint", side_effect=[7, 35, 60, 8, 4])
    def test_comercial_espacios_ingresos_gastos(self, _rnd):
        # tienda
        c1 = Comercial("C1", "Comercial", "tienda", comunitario=False)
        self.assertEqual(c1.altura, 15)
        self.assertEqual(c1.espacio, 7)  # valor asignado en __init__
        self.assertEqual(c1.calcular_ingresos(), 1500)
        self.assertEqual(c1.calcular_gastos(), 1000)

        # restaurante
        c2 = Comercial("C2", "Comercial", "restaurante", comunitario=True)
        self.assertEqual(c2.espacio, 35)
        self.assertEqual(c2.calcular_ingresos(), 30000)
        self.assertEqual(c2.calcular_gastos(), 18000)

        # supermercado
        c3 = Comercial("C3", "Comercial", "supermercado", comunitario=False)
        self.assertEqual(c3.espacio, 60)
        self.assertEqual(c3.calcular_ingresos(), 7000)
        self.assertEqual(c3.calcular_gastos(), 6000)

        # farmacia
        c4 = Comercial("C4", "Comercial", "farmacia", comunitario=True)
        self.assertEqual(c4.espacio, 8)
        self.assertEqual(c4.calcular_ingresos(), 2000)
        self.assertEqual(c4.calcular_gastos(), 1500)

        # libreria
        c5 = Comercial("C5", "Comercial", "libreria", comunitario=False)
        self.assertEqual(c5.espacio, 4)
        self.assertEqual(c5.calcular_ingresos(), 2000)
        self.assertEqual(c5.calcular_gastos(), 2000)

    def test_comercial_categoria_invalida(self):
        with self.assertRaises(ValueError):
            Comercial("Bad", "Residencial", "tienda", False)

    def test_comercial_tipo_invalido(self):
        with self.assertRaises(ValueError):
            Comercial("Bad", "Comercial", "kiosko espacial", False)



# Tests para las clases Personas

class TestPersonas(unittest.TestCase):
    def test_residente(self):
        p = Residente("Ana", "F")
        self.assertEqual(p.tipo, "residente")
        self.assertEqual(p.espacio, 1)

    def test_trabajador(self):
        p = Trabajador("Ben", "M")
        self.assertEqual(p.tipo, "trabajador")
        self.assertEqual(p.espacio, 3)

    def test_visitante(self):
        p = Visitante("Caro", "F")
        self.assertEqual(p.tipo, "visitante")
        self.assertEqual(p.espacio, 2)



# Tests para las clases Vehículos

class TestVehiculos(unittest.TestCase):
    def test_auto_por_tipo_propietario(self):
        a1 = Auto("a", "m", "mod", 2020, Residente("R","F"))
        self.assertEqual(a1.ingreso, 1000)
        self.assertEqual(a1.gasto, 400)

        a2 = Auto("a", "m", "mod", 2020, Trabajador("T","M"))
        self.assertEqual(a2.ingreso, 500)
        self.assertEqual(a2.gasto, 250)

        a3 = Auto("a", "m", "mod", 2020, Visitante("V","F"))
        self.assertEqual(a3.ingreso, 3000)
        self.assertEqual(a3.gasto, 100)

    def test_camioneta_por_tipo_propietario(self):
        v1 = Camioneta("c", "m", "mod", 2020, Residente("R","F"))
        self.assertEqual(v1.ingreso, 1200)
        self.assertEqual(v1.gasto, 420)

        v2 = Camioneta("c", "m", "mod", 2020, Trabajador("T","M"))
        self.assertEqual(v2.ingreso, 550)
        self.assertEqual(v2.gasto, 270)

        v3 = Camioneta("c", "m", "mod", 2020, Visitante("V","F"))
        self.assertEqual(v3.ingreso, 3500)
        self.assertEqual(v3.gasto, 120)

    def test_moto_por_tipo_propietario(self):
        m1 = Moto("m", "m", "mod", 2020, Residente("R","F"))
        self.assertEqual(m1.ingreso, 700)
        self.assertEqual(m1.gasto, 150)

        m2 = Moto("m", "m", "mod", 2020, Trabajador("T","M"))
        self.assertEqual(m2.ingreso, 350)
        self.assertEqual(m2.gasto, 100)

        m3 = Moto("m", "m", "mod", 2020, Visitante("V","F"))
        self.assertEqual(m3.ingreso, 2500)
        self.assertEqual(m3.gasto, 50)

    def test_camion_por_tipo_propietario(self):
        c1 = Camion("c", "m", "mod", 2020, Residente("R","F"))
        self.assertEqual(c1.ingreso, 2000)
        self.assertEqual(c1.gasto, 800)

        c2 = Camion("c", "m", "mod", 2020, Trabajador("T","M"))
        self.assertEqual(c2.ingreso, 1200)
        self.assertEqual(c2.gasto, 400)

        c3 = Camion("c", "m", "mod", 2020, Visitante("V","F"))
        self.assertEqual(c3.ingreso, 7000)
        self.assertEqual(c3.gasto, 300)

    def test_vehiculo_propietario_invalido(self):
        with self.assertRaises(ValueError):
            Auto("a","m","mod",2020, PruebaPersona(tipo="x")).calcular_ingreso()
        with self.assertRaises(ValueError):
            Camioneta("a","m","mod",2020, PruebaPersona(tipo="x")).calcular_gasto()
        with self.assertRaises(ValueError):
            Moto("a","m","mod",2020, PruebaPersona(tipo="x")).calcular_ingreso()
        with self.assertRaises(ValueError):
            Camion("a","m","mod",2020, PruebaPersona(tipo="x")).calcular_gasto()



# Tests para Torre

class TestTorre(unittest.TestCase):
    
    # Parcheamos __main__.randint, que es reemplazar dicha función (tambien se puede con atributos) de manera temporal durante un test. Para eso usamos @patch, que es una
    # herramienta de la librería de mock, de unittest "unittest.mock", para evitar aleatoriedad y hacer los tests deterministas y reproducibles, vale decir, que siempre 
    # da el mismo resultado si no cambias el código, y además que cualquiera, en cualquier computador, pueda ejecutar el test y obtener el mismo resultado.

    # En este caso, si hubiesemos importado randint de otra manera, tendríamos que cambiar la ruta en el patch, vale decir, si hubiesemos hecho:
    # import random -> y luego en el codigo usar randint de la forma: random.randint(), el parcheo hubieso incluido a random, vale decir: 
    # @patch("random.randint", side_effect=[41, 203]). Por otro lado, side_effect nos permite definir el comportamiento de la función parcheada. En este caso
    # que es un randint(), el primero valor de la lista será el primer numero que entregará la función  cuando se use, y el segundo el siguiente. Por lo tanto, 
    # si queremos usar el randint(1,100) para que me entregue un valor al azar entre 1 y 100, por el parcheo entregará primero 41, y después 203. 
    
    #Por ultimo, este parcheo que les expliqué, hace efecto en todas las llamadas a randint dentro del código que se ejecute en ese test.
    #Por ejemplo, en esta función "test_torre_atributos_basicos_y_calculos(self, _rnd)" hará efecto al instanciar la clase, Residencial, por el hecho de que es una clase
    #donde se utiliza randint, y así sucesivamente con las demas instancias. 

    @patch("__main__.randint", side_effect=[41, 203])
    def test_torre_atributos_basicos_y_calculos(self, _rnd): #Cuando se usa @patch, la función en el que incide recibe un argumento extra, que es el "mock" o el parcheo 
                                                            #que reemplazó a, en este caso, la función randint. _rnd es el objeto mock necesario para que se haga el reemplazo.
        espacios = {
            "Residencial": [
                Residencial("R1", "Residencial", "departamento pequeño"),
                Residencial("R2", "Residencial", "departamento mediano"),
            ],
            "Oficina": [
                Oficina("O1", "Oficina", "oficina"),
                Oficina("O2", "Oficina", "oficina de reuniones"),
            ],
        }
        estacionamiento = [PruebaVehiculo(100, 40), PruebaVehiculo(200, 50)]
        t = Torre("Gran Torre", espacios, estacionamiento)
        self.assertEqual(t.pisos, 45)
        self.assertEqual(t.capacidad_estacionamiento, 205)
        self.assertEqual(t.altura, 30)
        self.assertEqual(t.espacios_por_piso, 4 // 45)
        self.assertEqual(t.ingresos_estacionamiento, 300)
        self.assertEqual(t.gastos_estacionamiento, 90)
        self.assertEqual(t.utilidades, 210)
        self.assertAlmostEqual(t.ocupacion_estacionamiento(), 2 / 205) #assertAlmostEqual(a, b, places=7) --> Verifica que a y b sean casi iguales, 
                                                                       #permitiendo una pequeña diferencia (tolerancia). Por defecto compara hasta 7 decimales.
        s = str(t)
        self.assertIn("Torre Gran Torre", s) #assertIn(a, b) --> Verifica que a esté contenido en b. Sirve para listas, tuplas, strings, sets, etc.
        self.assertIn("45 pisos", s)
        self.assertIn("Capacidad de estacionamiento: 205", s)

    @patch("__main__.randint", side_effect=[40, 200])
    def test_multiplos_de_cinco(self, _rnd):
        t = Torre("T", {}, [])
        self.assertEqual(t.multiplos_de_cinco(41), 45)
        self.assertEqual(t.multiplos_de_cinco(40), 40)
        self.assertEqual(t.multiplos_de_cinco(44), 45)
        self.assertEqual(t.multiplos_de_cinco(0), 0)
        self.assertEqual(t.multiplos_de_cinco(300), 300)

    @patch("__main__.randint", side_effect=[40, 200])
    def test_ocupacion_capacidad_cero(self, _rnd):
        t = Torre("T", {}, [])
        t.capacidad_estacionamiento = 0
        self.assertEqual(t.ocupacion_estacionamiento(), 0)

    @patch("__main__.randint", side_effect=[50, 250])
    def test_calcular_espacios_por_piso(self, _rnd):
        espacios = {
            "Residencial": [
                Residencial("R1", "Residencial", "departamento pequeño"),
                Residencial("R2", "Residencial", "departamento mediano"),
                Residencial("R3", "Residencial", "departamento pequeño"),
            ],
            "Oficina": [
                Oficina("O1", "Oficina", "oficina"),
                Oficina("O2", "Oficina", "oficina de reuniones"),
                Oficina("O3", "Oficina", "oficina"),
            ],
        }
        t = Torre("T", espacios, [])
        self.assertEqual(t.espacios_por_piso, 6 // 50)

    @patch("__main__.randint", side_effect=[60, 200, 7]) 
    def test_calculo_altura_con_tipos(self, _rnd):
        espacios = {
            "Residencial": [Residencial("R", "Residencial", "departamento pequeño")],
            "Oficina": [Oficina("O", "Oficina", "oficina")],
            "Comercial": [Comercial("C", "Comercial", "tienda", comunitario=False)],
        }
        t = Torre("T", espacios, [])
        self.assertEqual(t.altura, 30)

    @patch("__main__.randint", side_effect=[45, 200])
    def test_ingresos_y_gastos_estacionamiento(self, _rnd):
        est = [PruebaVehiculo(10, 1), PruebaVehiculo(20, 2), PruebaVehiculo(30, 3)]
        t = Torre("T", {}, est)
        self.assertEqual(t.ingresos_estacionamiento, 60)
        self.assertEqual(t.gastos_estacionamiento, 6)
        self.assertEqual(t.utilidades, 54)


if __name__ == "__main__":
    unittest.main(argv=['first-arg-is-ignored'], exit=False, verbosity=2) #Esta linea es util para archivos jupyter notebook. En caso de que se quiera ejecutar las pruebas 
                                                                          #desde un archivo .py, solo basta con poner el parentesís. 
```
