# Estructuras de Datos en Python

## Estructuras Basicas

### Listas

Las listas son estructuras de datos mutables, lo que significa que puedes modificar su contenido. Son versátiles y permiten almacenar elementos de cualquier tipo, en un orden específico.

```python
frutas = ['manzana', 'banana', 'cereza']
frutas.append('durazno')
print(frutas)  # Output: ['manzana', 'banana', 'cereza', 'durazno']

# Acceso y modificación
frutas[1] = 'kiwi'
print(frutas)  # Output: ['manzana', 'kiwi', 'cereza', 'durazno']
```

### Tuplas

Las tuplas son estructuras inmutables, lo que significa que no pueden cambiarse después de ser creadas. Son útiles cuando necesitas una colección de elementos que no deben modificarse.

```python
punto = (3, 5)
print(punto[0], punto[1])  # Output: 3 5

# Desempaquetado de tupla
x, y = punto
print(f'Coordenadas: {x}, {y}')  # Output: Coordenadas: 3, 5
```

## Diccionarios

Los diccionarios en Python son estructuras de datos que permiten almacenar pares de elementos relacionados mediante una clave (key) y un valor (value). Esto permite un acceso eficiente a los valores utilizando las claves como referencia.

```python
# Creación de un diccionario
perros = {'bc': 'border-collie', 'lr': 'labrador retriever', 'pg': 'pug'}
print(perros)  # Output: {'bc': 'border-collie', 'lr': 'labrador retriever', 'pg': 'pug'}
```

```python
# Acceso a un valor
print(perros['bc'])  # Output: border-collie

# Modificación de un valor
perros['pg'] = 'pug-modificado'
print(perros['pg'])  # Output: pug-modificado

# Añadir un nuevo par clave-valor
perros['te'] = 'terrier'
print(perros)  # Output: {'bc': 'border-collie', 'lr': 'labrador retriever', 'pg': 'pug-modificado', 'te': 'terrier'}
```

**Algunos metodos utiles...**

```python
monedas = {'chile': 'peso', 'brasil': 'real', 'peru': 'sol'}

# Obtener las claves
print(monedas.keys())  # Output: dict_keys(['chile', 'brasil', 'peru'])

# Obtener los valores
print(monedas.values())  # Output: dict_values(['peso', 'real', 'sol'])

# Obtener pares clave-valor
print(monedas.items())  # Output: dict_items([('chile', 'peso'), ('brasil', 'real'), ('peru', 'sol')])
```


**Uso de args y kwargs**

Los *args y **kwargs permiten manejar argumentos variables en funciones. *args se usa para pasar una lista de argumentos posicionales, mientras que **kwargs se utiliza para pasar un diccionario de argumentos con nombres.

## Arboles

Un árbol es una estructura de datos no lineal que organiza los elementos en una jerarquía, donde cada elemento se llama nodo. El nodo principal o de más alto nivel se denomina raíz. A partir de la raíz, cada nodo puede tener cero o más hijos, y cada nodo hijo tiene exactamente un padre. Los nodos que comparten el mismo padre se llaman hermanos. Esta estructura es útil para representar jerarquías o relaciones anidadas.

```python
class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.hijos = []

    def agregar_hijo(self, nodo_hijo):
        self.hijos.append(nodo_hijo)

# Crear nodos
raiz = Nodo('Reino Animal')
vertebrados = Nodo('Vertebrados')
invertebrados = Nodo('Invertebrados')
gusanos = Nodo('Gusanos')

# Establecer relaciones jerárquicas
raiz.agregar_hijo(vertebrados)
raiz.agregar_hijo(invertebrados)
invertebrados.agregar_hijo(gusanos)

# Ejemplo de recorrido simple
def recorrer_arbol(nodo, nivel=0):
    print('  ' * nivel + nodo.valor)
    for hijo in nodo.hijos:
        recorrer_arbol(hijo, nivel + 1)

recorrer_arbol(raiz)
```

# Programación Orientada a Objetos

La Programación Orientada a Objetos (OOP) es un paradigma de programación que utiliza objetos y sus interacciones para diseñar y desarrollar aplicaciones. Los objetos en programación son representaciones virtuales de entidades del mundo real o conceptos, que tienen atributos (datos) y métodos (comportamientos).

*Conceptos Clave de OOP*

**Clase**: Es un molde o plantilla para crear objetos. Define los atributos y métodos que los objetos creados a partir de la clase tendrán.

**Objeto**: Es una instancia de una clase. Los objetos son entidades que contienen atributos y métodos definidos en la clase.

**Encapsulación**: Es la práctica de restringir el acceso directo a algunos componentes de un objeto, protegiendo su estado interno y exponiendo solo lo necesario a través de una interfaz.

**Interfaz**: Es un conjunto de métodos accesibles que permiten interactuar con el objeto. La interfaz es la "fachada" que oculta la complejidad interna.

Definición de una clase

```python
class Vehiculo:
    def __init__(self, marca, modelo, year):
        self.marca = marca
        self.modelo = modelo
        self.year = year
        self.encendido = False

    def encender(self):
        if not self.encendido:
            self.encendido = True
            print(f"El {self.marca} {self.modelo} está encendido.")
        else:
            print(f"El {self.marca} {self.modelo} ya está encendido.")

    def apagar(self):
        if self.encendido:
            self.encendido = False
            print(f"El {self.marca} {self.modelo} está apagado.")
        else:
            print(f"El {self.marca} {self.modelo} ya está apagado.")
```

Creación de un objeto

```python
auto1 = Vehiculo("Toyota", "Corolla", 2020)
auto2 = Vehiculo("Honda", "Civic", 2019)

auto1.encender()  # Output: El Toyota Corolla está encendido.
auto2.encender()  # Output: El Honda Civic está encendido.

auto1.apagar()    # Output: El Toyota Corolla está apagado.
```

## Herencia

La herencia es un concepto fundamental en la programación orientada a objetos que permite que una clase derive características y comportamientos de otra clase. Esto nos permite reutilizar código y representar relaciones jerárquicas entre clases.

*Conceptos Básicos*

**Clase Padre (Superclase):** Es la clase de la cual otras clases pueden heredar atributos y métodos.

**Clase Hija (Subclase):** Es una clase que hereda de otra clase. Puede agregar nuevos atributos y métodos, o sobrescribir los existentes en la superclase.

```python
# Clase Padre
class Vehiculo:
    def __init__(self, marca, modelo, año):
        self.marca = marca
        self.modelo = modelo
        self.año = año

    def describir(self):
        return f"Vehículo: {self.marca} {self.modelo} ({self.año})"

# Clase Hija
class FurgonEscolar(Vehiculo):
    def __init__(self, marca, modelo, año, lista_ninos):
        super().__init__(marca, modelo, año)  # Llamada al constructor de la superclase
        self.lista_ninos = lista_ninos

    def describir(self):
        descripcion_base = super().describir()  # Llamada al método de la superclase
        return f"{descripcion_base}, transporta a {len(self.lista_ninos)} niños."

# Ejemplo de uso
furgon = FurgonEscolar("Ford", "Transit", 2020, ["Juan", "María", "Pedro"])
print(furgon.describir())  # Output: Vehículo: Ford Transit (2020), transporta a 3 niños.
```

## Polimorfismo

El polimorfismo permite que diferentes clases respondan al mismo método de diferentes maneras. Esto es útil cuando se quiere que varias clases compartan una interfaz común pero tengan implementaciones específicas.

```python
class Animal:
    def sonido(self):
        raise NotImplementedError("Este método debe ser implementado por una subclase")

class Perro(Animal):
    def sonido(self):
        return "Guau"

class Gato(Animal):
    def sonido(self):
        return "Miau"

# Ejemplo de polimorfismo
animales = [Perro(), Gato()]

for animal in animales:
    print(animal.sonido())  # Output: Guau Miau
```

## Sobrecarga

En Python, podemos sobrecargar operadores para que funcionen de manera específica con nuestras clases. Por ejemplo, podemos definir cómo se comporta el operador + cuando se utiliza con objetos de una clase personalizada.

```python
class Carro:
    def __init__(self, productos):
        self.productos = productos

    def __add__(self, otro):
        resultado = self.productos.copy()
        for producto, cantidad in otro.productos.items():
            if producto in resultado:
                resultado[producto] += cantidad
            else:
                resultado[producto] = cantidad
        return Carro(resultado)

    def __str__(self):
        return str(self.productos)

# Ejemplo de uso
carro1 = Carro({'pan': 2, 'leche': 1})
carro2 = Carro({'leche': 3, 'huevos': 12})
carro3 = carro1 + carro2
print(carro3)  # Output: {'pan': 2, 'leche': 4, 'huevos': 12}
```

## Multiherencia en Python


La multiherencia en Python permite que una subclase herede atributos y métodos de más de una superclase. Esto puede ser útil para crear clases que combinan funcionalidades de diferentes fuentes. Sin embargo, la multiherencia puede llevar a situaciones complicadas, como el problema del diamante, donde la jerarquía de clases forma una estructura en la que una superclase común se hereda más de una vez.



```python
class Investigador:
    def __init__(self, area):
        self.area = area

class Docente:
    def __init__(self, departamento):
        self.departamento = departamento

class Academico(Docente, Investigador):
    def __init__(self, nombre, area_investigacion, departamento):
        Investigador.__init__(self, area_investigacion)
        Docente.__init__(self, departamento)
        self.nombre = nombre

# Creación de un objeto de la clase Academico
p1 = Academico("Juan Perez", "Inteligencia de Máquina", "Ciencia De La Computación")
print(p1.nombre)        # Output: Juan Perez
print(p1.area)          # Output: Inteligencia de Máquina
print(p1.departamento)  # Output: Ciencia De La Computación
```

### Problema del Diamante


El problema del diamante ocurre cuando una clase hereda de dos clases que, a su vez, heredan de una misma clase base. Esto puede llevar a que métodos de la clase base sean llamados más de una vez si no se maneja correctamente.

```python
class ClaseB:
    num_llamadas_B = 0
    def llamar(self):
        print("Llamando método en Clase B")
        self.num_llamadas_B += 1

class SubClaseIzquierda(ClaseB):
    num_llamadas_izq = 0
    def llamar(self):
        ClaseB.llamar(self)
        print("Llamando método en Subclase Izquierda")
        self.num_llamadas_izq += 1

class SubClaseDerecha(ClaseB):
    num_llamadas_der = 0
    def llamar(self):
        ClaseB.llamar(self)
        print("Llamando método en Subclase Derecha")
        self.num_llamadas_der += 1

class SubClaseA(SubClaseIzquierda, SubClaseDerecha):
    num_llamadas_subA = 0
    def llamar(self):
        SubClaseIzquierda.llamar(self)
        SubClaseDerecha.llamar(self)
        print("Llamando método en Subclase A")
        self.num_llamadas_subA += 1

s = SubClaseA()
s.llamar()
print(s.num_llamadas_subA, s.num_llamadas_izq, s.num_llamadas_der, s.num_llamadas_B)
```

Aquí, el** método llamar **de la clase ClaseB **se ejecuta dos veces** porque ambas subclases (SubClaseIzquierda y SubClaseDerecha) llaman a su método llamar.

Para evitar que el método de la clase base se ejecute más de una vez, Python utiliza el método super() y el Orden de Resolución de Métodos (MRO, por sus siglas en inglés) para manejar la jerarquía de llamadas de manera adecuada.

```python
class ClaseB:
    num_llamadas_B = 0
    def llamar(self):
        print("Llamando método en Clase B")
        self.num_llamadas_B += 1

class SubClaseIzquierda(ClaseB):
    num_llamadas_izq = 0
    def llamar(self):
        super().llamar()
        print("Llamando método en Subclase Izquierda")
        self.num_llamadas_izq += 1

class SubClaseDerecha(ClaseB):
    num_llamadas_der = 0
    def llamar(self):
        super().llamar()
        print("Llamando método en Subclase Derecha")
        self.num_llamadas_der += 1

class SubClaseA(SubClaseIzquierda, SubClaseDerecha):
    num_llamadas_subA = 0
    def llamar(self):
        super().llamar()
        print("Llamando método en Subclase A")
        self.num_llamadas_subA += 1

s = SubClaseA()
s.llamar()
print(s.num_llamadas_subA, s.num_llamadas_izq, s.num_llamadas_der, s.num_llamadas_B)
```

El MRO define el orden en el que se buscan los métodos en la jerarquía de clases en Python. Esto garantiza que cada método de la cadena de herencia se llame una sola vez en el orden correcto.

## Clases Abstractas

Una clase abstracta es aquella que no puede ser instanciada por sí sola. En lugar de ello, su propósito es ser heredada por otras clases que implementen sus métodos abstractos. Los métodos abstractos son aquellos que se definen en la clase abstracta, pero que no tienen una implementación concreta, **forzando a las subclases a implementarlos.**

```python
from abc import ABC, abstractmethod

class Vehiculo(ABC):
    @abstractmethod
    def conducir(self):
        pass

    @abstractmethod
    def capacidad_pasajeros(self):
        pass

class Auto(Vehiculo):
    def conducir(self):
        print("Conduciendo un auto")

    def capacidad_pasajeros(self):
        return 5

class Camion(Vehiculo):
    def conducir(self):
        print("Conduciendo un camión")

    def capacidad_pasajeros(self):
        return 2

# Ejemplo de uso
#v = Vehiculo()  # Esto generaría un error, ya que Vehiculo es una clase abstracta
auto = Auto()
camion = Camion()

auto.conducir()  # Output: Conduciendo un auto
print(auto.capacidad_pasajeros())  # Output: 5

camion.conducir()  # Output: Conduciendo un camión
print(camion.capacidad_pasajeros())  # Output: 2
```
