# Modelos predictivos de aprendizaje no supervisado

En la primera parte de este capítulo estudiamos modelos de aprendizaje supervisado, donde cada observación viene acompañada de una variable objetivo que queremos predecir. En este notebook estudiaremos una familia distinta de técnicas: el **aprendizaje no supervisado**.

En aprendizaje no supervisado no buscamos entrenar un modelo para predecir una etiqueta conocida. En cambio, buscamos descubrir estructura en los datos. Esto puede significar encontrar grupos de observaciones similares, reducir la dimensionalidad de un conjunto de datos, visualizar patrones complejos o identificar comportamientos que no eran evidentes a simple vista.

En este notebook trabajaremos con un conjunto de datos asociado a eficiencia energética de edificios. El objetivo será analizar si existen grupos de edificios con características similares y estudiar cómo esos grupos se relacionan con sus cargas de calefacción y enfriamiento.

## ¿Qué es el aprendizaje no supervisado?

En aprendizaje supervisado, cada dato tiene asociada una respuesta conocida. Por ejemplo, podríamos tener información de edificios junto con su consumo energético, y entrenar un modelo para predecir ese consumo a partir de sus características.

En aprendizaje no supervisado, en cambio, el modelo no recibe una respuesta correcta para cada ejemplo. Solo recibe las características de los datos. Por lo tanto, la pregunta cambia: ya no buscamos predecir una variable objetivo, sino encontrar patrones internos.

Algunas preguntas típicas de aprendizaje no supervisado son:

- ¿Existen grupos naturales dentro de los datos?
- ¿Qué observaciones son parecidas entre sí?
- ¿Podemos representar los datos usando menos variables?
- ¿Qué variables parecen estar relacionadas con la estructura de los datos?
- ¿Los grupos encontrados tienen una interpretación útil desde el punto de vista del problema?

En este notebook nos centraremos en dos tareas:

1. **Reducción de dimensionalidad**: representar datos con muchas variables usando menos dimensiones.
2. **Clustering**: agrupar observaciones similares sin conocer previamente las categorías.

## Técnicas de aprendizaje no supervisado

Típicamente, las técnicas de aprendizaje no supervisado se asocian con dos tareas, el _clustering_ y la reducción de dimensionalidad. A continuación, se describirán las características de algunos de los algoritmos típicos para enfrentar estas tareas.

### Algoritmos de clustering
Se utilizan principalmente para dividir un conjunto de datos en grupos (llamados ``clústers'') de forma tal que los elementos de un clúster sean similares entre sí. Las técnicas de clustering son útiles en muchos contextos, como la segmentación de clientes o la agrupación de documentos por tema. Los dos esquemas de clustering más comunes son el esquema jerárquico y el de centroide.

#### Esquemas jerárquicos
Los esquemas jerárquicos se basan en la construcción de una estructura jerárquica de relación entre los ejemplos, en base a la idea de que ejemplos cercanos en el espacio de características están más relacionados entre sí que los que están más separados. El algoritmo de clustering jerárquico más importante es el Agglomerative Hierarchical Clustering, o AHC, que trabaja uniendo puntos uno a uno en base a su distancia. Este algoritmo es muy popular debido a su simplicidad de funcionamiento e implementación, además de que no requiere de la especificación previa de la cantidad de clústers que se generarán. Por otro lado, el algoritmo presenta dificultades para definir el umbral de corte preciso, lo que además dificulta el manejo de outliers, ya que pueden incluirse fácilmente en un clúster ya existente si no se considera un umbral especial para ellos.

<img src="figs/ahc.png" width="600px" align="center"/>

#### Esquemas de centroide
Los esquemas de centroide se basan en la idea de que cada clúster es representado por un vector de características, que actúa como el centroide o la media de los elementos que lo componen. Luego, la pertenencia un ejemplo a un clúster se define en base a la distancia con respecto a este centroide en el espacio de características. El algoritmo más importante para este modelo de clúster, y quizá el algoritmo de clustering más popular en general, es K-means. Este algoritmo divide el conjunto de datos en un número predefinido (K) de clústers, usando típicamente como métrica de distancia a la distancia euclidiana. El representante de cada grupo se llama centroide y corresponde al vector de características promedio de los ejemplos asignados a ese clúster. Si bien funcionamiento simple e intuitivo de K-means hace que sea muy popular, tiene la gran desventaja de que es necesario seleccionar por adelantado el número K de clústers. Para llevar a cabo esto, se utiliza típicamente el método _Silhouette_ o el método del codo.

### Reducción de dimensionalidad
Independiente del paradigma de aprendizaje que se utilice, en Machine Learning los datos forman la base de cualquier algoritmo de aprendizaje. Por este motivo, las características que se utilizan para modelar los datos idealmente deben capturar los atributos fundamentales de estos, con el fin de facilitar el trabajo de los algoritmos. Sin embargo, muchas veces el número de características resulta ser excesivo, ya sea debido a redundancia o que algunas de las cuales derechamente ni siquiera son afines a los datos. Este exceso de características puede no solo complicar a los modelos de aprendizaje con respecto a su posibilidad de sobreajuste, sino que además hacen más difícil, la interpretación y comprensión de los datos y los resultados a través de su visualización. Aquí es donde entran en juego las técnicas de reducción de la dimensionalidad.

La reducción de dimensionalidad es la tarea de reducir la cantidad de características en un conjunto de datos. El proceso de reducción esencialmente transforma los datos de un espacio de características de alta dimensión a uno de baja dimensión, cuidando que las propiedades más significativas presentes en los datos no se pierdan durante la transformación.

Es importante destacar que la tarea de reducción de dimensionalidad es distinta de la de selección de características. Si bien ambas generan como resultado un número de características menor al original, la selección de características elige aquellas que mejor permiten resolver una tarea predictiva en particular, mientras que la de reducción crea nuevas características a partir de las originales, en base a los atributos intrínsecos de los datos, no necesitando así la definición explícita de una tarea de predicción.

<img src="figs/dimensionality_reduction.png" width="400px" align="center"/>

Existen dos familias principales de algoritmos de reducción de dimensionalidad, los de proyección lineal y los de aprendizaje de manifold. A continuación, describiremos ambas familias, junto con sus algoritmos más representativos.

#### Algoritmos de proyección lineal
Como su nombre lo indica, estos algoritmos transforman los datos a través de la aplicación de operaciones lineales, de modo que la proyección cumpla ciertas condiciones. El algoritmo más utilizado de esta familia es el _Principal Component Analysis_ o PCA, el cual busca proyectar los datos a un espacio de menor dimensionalidad, preservando la varianza presente en ellos. Para esto, es necesario buscar los componentes principales de los datos, que corresponden a los vectores y valores propios de la matriz de características de estos. Estos vectores definirán los ejes del nuevo espacio de características, donde aquellos con mayores valores propios capturarán la mayor cantidad de varianza de los datos.

De esta forma, es posible reducir la dimensionalidad proyectando los datos a un espacio con una cantidad de ejes menor a la original, multiplicando la matriz de características por los una fracción de los vectores propios encontrados, en orden de mayor a menor captura de varianza. Es decir, la primera componente principal captura la mayor cantidad posible de variabilidad de los datos, la segunda componente captura la mayor cantidad posible de la variabilidad restante, manteniéndose independiente de la primera, y así sucesicamente. Al igual que todos los algoritmos de esta familia, PCA tiene problemas para proyectar datos cuyas características presenten relaciones no lineales.

#### Algoritmos de aprendizaje de manifold
Los algoritmos de aprendizaje de manifold evitan las limitantes de los algoritmos de proyección lineal, al utilizar transformaciones no lineales para mover los datos del espacio de origen a un manifold de destino, que mantiene las propiedades locales de los ejemplos. Uno de los algoritmos de esta familia más utilizados es el _t-Distributed Stochastic Neighbor Embedding_ o t-SNE, que es muy adecuada además para la visualización de datos. A diferencia de PCA, que simplemente maximiza la captura de la varianza de los datos, t-SNE minimiza la divergencia entre distribuciones de datos. Esencialmente, captura localmente la distribución de los en el espacio de alta dimensión y luego aprende una proyección no lineal que mantenga lo más similar posible esta distribución, pero en el espacio de baja dimensionalidad. Esto último se logra a través del aprendizaje de una proyección no lineal, mediante un problema de minimización, donde la función objetivo corresponde a la divergencia KL. Un detalle importante de los algoritmos de esta familia es su gran demanda por recursos computacionales, lo que lleva a utilizar estrategias combinadas cuando los espacios de origen son de muy alta dimensionalidad. Específicamente, se recomienda primero aplicar PCA para llevar los datos a un espacio de menor dimensionalidad, menor o igual a 50 dimensiones por ejemplo, y luego aplicar t-SNE.

Para concluir, podemos realizar una comparación visual de estos algoritmos, para hacernos una idea de sus diferencias:

Al reducir a 2 dimensiones el espacio de características del conjunto de datos MNIST, que consiste en imágenes de 28x28 pixeles de dígitos entre 0 y 9 escritos a manos, podemos notar claras diferencias en el resultado. Es posible apreciar que si bien todos los algoritmos capturan grupos que coinciden con las etiquetas, los resultados de t-SNE son mucho más satisfactorios, separando de mejor manera los puntos.

## Dataset: eficiencia energética de edificios

Utilizaremos el dataset **Energy Efficiency**, disponible en UCI Machine Learning Repository. Este dataset contiene simulaciones de edificios con distintas características geométricas y constructivas. Para cada edificio se dispone de variables como compacidad relativa, área de superficie, área de muros, área de techo, altura, orientación, área vidriada y distribución del área vidriada.

El dataset también contiene dos variables de respuesta:

- carga de calefacción;
- carga de enfriamiento.

Como este notebook trata sobre aprendizaje no supervisado, estas dos variables no serán usadas para entrenar los modelos. Las usaremos solamente al final, para interpretar los grupos encontrados.

## Importación de librerías

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score

%matplotlib inline
```

## Carga de datos

El archivo original se encuentra disponible en formato Excel. Lo cargaremos directamente desde el repositorio de UCI.

Cada fila representa una configuración de edificio. Las columnas `Y1` e `Y2` corresponden a las cargas de calefacción y enfriamiento, respectivamente.

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00242/ENB2012_data.xlsx"

energy_dataset = pd.read_excel(url)
energy_dataset.head()
```

```python
energy_dataset.info()
```

Los nombres originales de las columnas son poco descriptivos. Para facilitar la interpretación, cambiaremos los nombres por otros más informativos.

```python
energy_dataset = energy_dataset.rename(columns={
    "X1": "relative_compactness",
    "X2": "surface_area",
    "X3": "wall_area",
    "X4": "roof_area",
    "X5": "overall_height",
    "X6": "orientation",
    "X7": "glazing_area",
    "X8": "glazing_area_distribution",
    "Y1": "heating_load",
    "Y2": "cooling_load"
})

energy_dataset.head()
```

```python
energy_dataset.describe()
```

## Matriz de características

Para aplicar aprendizaje no supervisado construiremos una matriz `X` usando solo las variables que describen el diseño del edificio.

Las variables `heating_load` y `cooling_load` no se usarán para entrenar los modelos. Las guardaremos aparte porque más adelante nos ayudarán a interpretar si los grupos encontrados tienen sentido desde el punto de vista energético.

```python
features = [
    "relative_compactness",
    "surface_area",
    "wall_area",
    "roof_area",
    "overall_height",
    "orientation",
    "glazing_area",
    "glazing_area_distribution"
]

target_vars = ["heating_load", "cooling_load"]

X = energy_dataset[features].copy()
energy_targets = energy_dataset[target_vars].copy()

X.head()
```

## Escalamiento de variables

Muchas técnicas no supervisadas dependen de distancias entre observaciones. Por ejemplo, K-means asigna cada observación al centroide más cercano. Si una variable está medida en una escala mucho mayor que las demás, puede dominar artificialmente el cálculo de distancias.

Por esta razón, escalaremos las variables usando `StandardScaler`.

En este caso no separaremos los datos en entrenamiento y test, porque no estamos entrenando un modelo predictivo supervisado. El objetivo es analizar la estructura del conjunto completo de datos.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled[:5]
```

## Reducción de dimensionalidad con PCA

Reduciremos los datos desde 8 variables originales a 2 componentes principales, lo que nos permitirá visualizar los edificios en un plano.

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

pca_df = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
pca_df["heating_load"] = energy_dataset["heating_load"]
pca_df["cooling_load"] = energy_dataset["cooling_load"]

pca_df.head()
```

```python
print("Varianza explicada por cada componente:")
print(pca.explained_variance_ratio_)

print(f"Varianza total explicada por las dos primeras componentes: {pca.explained_variance_ratio_.sum():.2%}")
```

```python
plt.figure(figsize=(7, 5))
scatter = plt.scatter(pca_df["PC1"], pca_df["PC2"], c=pca_df["heating_load"])
plt.title("Representación PCA de edificios según carga de calefacción")
plt.xlabel("Primera componente principal")
plt.ylabel("Segunda componente principal")
plt.colorbar(scatter, label="Carga de calefacción")
plt.show()
```

```python
plt.figure(figsize=(7, 5))
scatter = plt.scatter(pca_df["PC1"], pca_df["PC2"], c=pca_df["cooling_load"])
plt.title("Representación PCA de edificios según carga de enfriamiento")
plt.xlabel("Primera componente principal")
plt.ylabel("Segunda componente principal")
plt.colorbar(scatter, label="Carga de enfriamiento")
plt.show()
```

Los gráficos anteriores usan las cargas de calefacción y enfriamiento solo como referencia visual. PCA no utilizó estas variables para construir las componentes principales.

Esto es importante, ya que estamos analizando si la estructura de las variables de diseño se relaciona con el desempeño energético de los edificios, pero no estamos entrenando un modelo para predecir ese desempeño.

## Interpretación de las componentes principales

PCA construye componentes como combinaciones de las variables originales. Para entender mejor qué representa cada componente, podemos analizar los pesos o cargas asociados a cada variable.

Estos valores no siempre tienen una interpretación directa, pero permiten identificar qué variables contribuyen más a cada componente.

```python
loadings = pd.DataFrame(
    pca.components_.T,
    columns=["PC1", "PC2"],
    index=features
)

loadings
```

```python
loadings.plot(kind="bar", figsize=(9, 5))
plt.title("Contribución de variables originales a las componentes principales")
plt.xlabel("Variable")
plt.ylabel("Peso en la componente")
plt.xticks(rotation=45, ha="right")
plt.show()
```

## Reducción de dimensionalidad con t-SNE

t-SNE puede producir visualizaciones útiles, pero debe interpretarse con cuidado. Las distancias globales entre grupos no siempre tienen una interpretación directa, y los resultados pueden cambiar al modificar sus hiperparámetros.

```python
tsne = TSNE(
    n_components=2,
    perplexity=30,
    learning_rate="auto",
    init="pca",
    random_state=0
)

X_tsne = tsne.fit_transform(X_scaled)

tsne_df = pd.DataFrame(X_tsne, columns=["Dim 1", "Dim 2"])
tsne_df["heating_load"] = energy_dataset["heating_load"]
tsne_df["cooling_load"] = energy_dataset["cooling_load"]

tsne_df.head()
```

```python
plt.figure(figsize=(7, 5))
scatter = plt.scatter(tsne_df["Dim 1"], tsne_df["Dim 2"], c=tsne_df["heating_load"])
plt.title("Representación t-SNE según carga de calefacción")
plt.xlabel("Dimensión 1")
plt.ylabel("Dimensión 2")
plt.colorbar(scatter, label="Carga de calefacción")
plt.show()
```

```python
plt.figure(figsize=(7, 5))
scatter = plt.scatter(tsne_df["Dim 1"], tsne_df["Dim 2"], c=tsne_df["cooling_load"])
plt.title("Representación t-SNE según carga de enfriamiento")
plt.xlabel("Dimensión 1")
plt.ylabel("Dimensión 2")
plt.colorbar(scatter, label="Carga de enfriamiento")
plt.show()
```

## Clustering con K-means

El algoritmo funciona de manera iterativa:

1. se inicializan `K` centroides;
2. cada punto se asigna al centroide más cercano;
3. se actualizan los centroides usando los puntos asignados a cada grupo;
4. se repite el proceso hasta que las asignaciones se estabilizan.

Una dificultad importante es que el número de clústers `K` debe definirse antes de ejecutar el algoritmo. Por eso, normalmente se prueban distintos valores de `K`.

```python
cluster_var = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init="k-means++", n_init=10, random_state=0)
    kmeans.fit(X_scaled)
    cluster_var.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(7, 5))
plt.plot(range(1, 11), cluster_var, marker="o")
plt.title("Método del codo")
plt.xlabel("Número de clústers K")
plt.ylabel("Inercia")
plt.xticks(range(1, 11))
plt.show()
```

La inercia mide qué tan cerca están los puntos de los centroides de sus respectivos clústers. Al aumentar `K`, la inercia siempre disminuye, porque cada clúster puede volverse más específico.

Por eso, no basta con elegir el valor de `K` que produce menor inercia. El método del codo busca identificar un punto a partir del cual aumentar `K` genera mejoras cada vez menores.

```python
silhouette_results = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, init="k-means++", n_init=10, random_state=0)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    silhouette_results.append({"k": k, "silhouette_score": score})

silhouette_df = pd.DataFrame(silhouette_results)
silhouette_df
```

```python
plt.figure(figsize=(7, 5))
plt.plot(silhouette_df["k"], silhouette_df["silhouette_score"], marker="o")
plt.title("Silhouette score para distintos valores de K")
plt.xlabel("Número de clústers K")
plt.ylabel("Silhouette score")
plt.xticks(range(2, 11))
plt.show()
```

El silhouette score compara qué tan cerca está cada observación de los elementos de su propio clúster en relación con los elementos de otros clústers. Valores más altos indican una separación más clara entre grupos.

A diferencia de la inercia, el silhouette score no necesariamente mejora al aumentar `K`, por lo que puede ayudar a elegir una cantidad razonable de clústers.

## Selección de un número de clústers

Para continuar con el análisis seleccionaremos automáticamente el valor de `K` que obtiene el mayor silhouette score dentro del rango evaluado.

Esta decisión no debe entenderse como una regla universal. En problemas reales, la selección del número de clústers también debe considerar la interpretación del problema, la estabilidad de los grupos y el uso que se dará a los resultados.

```python
best_k = int(silhouette_df.loc[silhouette_df["silhouette_score"].idxmax(), "k"])
best_score = silhouette_df["silhouette_score"].max()

print(f"Valor seleccionado de K: {best_k}")
print(f"Silhouette score asociado: {best_score:.3f}")
```

```python
kmeans = KMeans(n_clusters=best_k, init="k-means++", n_init=10, random_state=0)
kmeans_labels = kmeans.fit_predict(X_scaled)

pca_df["kmeans_cluster"] = kmeans_labels
energy_dataset["kmeans_cluster"] = kmeans_labels

pca_df.head()
```

```python
plt.figure(figsize=(7, 5))

for cluster in sorted(pca_df["kmeans_cluster"].unique()):
    subset = pca_df[pca_df["kmeans_cluster"] == cluster]
    plt.scatter(subset["PC1"], subset["PC2"], label=f"Clúster {cluster}")

plt.title("Clústers encontrados por K-means sobre visualización PCA")
plt.xlabel("Primera componente principal")
plt.ylabel("Segunda componente principal")
plt.legend()
plt.show()
```

El gráfico anterior muestra los clústers encontrados por K-means, proyectados sobre las dos primeras componentes principales.

Es importante notar que K-means fue entrenado usando las 8 variables escaladas, no las dos componentes del gráfico. La visualización sirve para interpretar el resultado, pero el clustering se realizó en el espacio completo de características.

```python
cluster_summary = energy_dataset.groupby("kmeans_cluster")[features + target_vars].mean()
cluster_summary
```

```python
cluster_summary[["heating_load", "cooling_load"]].plot(kind="bar", figsize=(8, 5))
plt.title("Carga promedio de calefacción y enfriamiento por clúster")
plt.xlabel("Clúster")
plt.ylabel("Carga promedio")
plt.xticks(rotation=0)
plt.show()
```

La tabla y el gráfico anteriores permiten interpretar los clústers desde el punto de vista del problema de ingeniería.

Aunque las cargas de calefacción y enfriamiento no se usaron para construir los clústers, podemos analizarlas después para estudiar si los grupos encontrados tienen diferencias relevantes en desempeño energético.

Este tipo de análisis es común en problemas no supervisados: primero se descubren grupos usando variables disponibles y luego se interpretan esos grupos usando conocimiento del dominio.

## Clustering jerárquico aglomerativo

En este notebook usaremos una versión simple, definiendo directamente el número final de clústers, con el fin de facilitar la interpretación.

```python
agg = AgglomerativeClustering(n_clusters=best_k)
agg_labels = agg.fit_predict(X_scaled)

pca_df["agg_cluster"] = agg_labels
energy_dataset["agg_cluster"] = agg_labels

pca_df.head()
```

```python
plt.figure(figsize=(7, 5))

for cluster in sorted(pca_df["agg_cluster"].unique()):
    subset = pca_df[pca_df["agg_cluster"] == cluster]
    plt.scatter(subset["PC1"], subset["PC2"], label=f"Clúster {cluster}")

plt.title("Clústers encontrados por clustering jerárquico")
plt.xlabel("Primera componente principal")
plt.ylabel("Segunda componente principal")
plt.legend()
plt.show()
```

```python
agg_summary = energy_dataset.groupby("agg_cluster")[features + target_vars].mean()
agg_summary
```

```python
agg_summary[["heating_load", "cooling_load"]].plot(kind="bar", figsize=(8, 5))
plt.title("Carga promedio por clúster jerárquico")
plt.xlabel("Clúster")
plt.ylabel("Carga promedio")
plt.xticks(rotation=0)
plt.show()
```

## Comparación de enfoques

K-means y clustering jerárquico pueden producir agrupaciones distintas, incluso cuando se aplican sobre los mismos datos. Esto ocurre porque cada algoritmo define de forma distinta qué significa construir buenos grupos.

K-means busca grupos compactos alrededor de centroides. Esto suele funcionar bien cuando los clústers tienen formas relativamente regulares y tamaños comparables.

El clustering jerárquico, en cambio, construye una secuencia de uniones entre observaciones o grupos. Esto permite capturar relaciones de agrupamiento de otra manera, aunque también depende de decisiones como la métrica de distancia y el criterio de enlace.

En problemas reales no existe necesariamente una única agrupación correcta. La calidad de un clustering depende tanto de criterios cuantitativos como de su utilidad para interpretar el problema.
